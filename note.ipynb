### 金融工程概述

#### 风险管理方法
+ 风险管理
    + 风险分散（投资组合理论）
    + 风险转移（金融工程）
        + 部分转移(期权)：有可能不行权，保留了部分不确定性
        + 全部转移（远期&期货）

#### 金融工程：本质&作用
+ 金融衍生产品的本质透析：
    + 零和博弈；
    + 契约性、虚拟性与未来性；
    + 高杠杆性；（反而可能造成更大的风险）
    + 衍生性
+ 金融衍生产品的功能：
    + 风险管理功能；
    + 价格发现功能；
    + 信息功能
+ 金融衍生产品的风险：如果监管和运用不当，则可能带来更大的风险

#### 金融产品定价理论
+ 1896 欧文·费雪 & 直接定价法：$PV = \Sigma \frac{CF_{t}}{(1+r)^{t}} $（如何确定贴现率r？）
+ 1938 麦考莱：“久期”、“利率免疫”
+ 1952 马科维茨： “证券组合分析” Mean-Var Analysis
+ 1960s 夏普：“资本资产定价模型”（CAPM）$E(r_{i})=r_{f}+β_{im}(E(r_{m})-r_{f})$ $β$衡量系统性风险
+ 1973 布莱克、舒尔斯 & 默顿：B-S期权定价模型
+ 1980s 达菲： 拓展到不完全竞争市场

#### 衍生品定价的基本假设
+ 市场不存在摩擦
+ 市场参与者不承担对手风险（Counterparty Risk） （OTC产品定价变简单）
+ 市场是完全竞争的
+ 市场不存在无风险套利机会，或者说市场参与者认为财富越多越好 （用于检验价格“合理性”）

#### 连续复利（一年计息无数次的利率）
先列出m次计息情况下的终值：
$$A(1+\frac{r_{m}}{m})^{mn}$$
利用高数里面的重要极限：
$$\lim_{x→\infty}(1+\frac{1}{x})^{x} = e$$
整理原式之后，代入极限：
$$(A(1+\frac{1}{\frac{m}{r_{m}}})^{\frac{m}{r_{m}}})^{r_{m}n}  $$
$$A(1+ e^{rn})$$
$A(1+ e^{rn})$表示连续复利情况下的终值

##### 各种计息频次利率与连续利率的转化公式
基础：令终值相等，列等式
$$e^{r} = (1+\frac{r_{m}}{m})^{m} = 1+r_{1}$$
即可推出：
$$r = m ln (1+\frac{r_{m}}{m}) $$

$$r_{m} = m(e^{\frac{r}{m}}-1)$$
尤其是当m = 1时： 
$$r = ln (1+r_{1}) = ln (P_{1})-ln (P_{0})$$

在股票投资场景中，由于对数收益率$ln (\frac{P_{1}}{P_{0}})$比百分比收益率$\frac{P_{1}-P_{0}}{P_{0}}$拥有正态性、可连加性等优良性质，一般对数收益率更加常用

### 远期&期货概述

#### 远期合约种类
1. 远期*利率*协议（FRA）
    + 定义：买卖双方同意从未来某一商定的时刻开始，在某一特定时期内按协议利率借贷一笔数额确定、以特定货币表示的名义本金的协议。
    + 通过事先约定利率，借款人可以规避利率上升的风险，贷款人则可以规避利率下跌的风险。无论未来利率升跌，FRA将实际贷款利率锁定为协议利率。
    + $1*4$：一个月后，借款3个月（至第四个月）
2. 远期*外汇*协议
    + 定义：双方约定在将来某一时间按约定的汇率买卖一定金额的某种外汇的合约
    + CNY:在岸市场的远期外汇交易，标的为在岸即期￥汇率；CNH：离岸市场的远期外汇交易，标的为离岸即期￥汇率；NDF:游离于监管之外的外汇市场，标的为在岸市场的￥汇率（CNY、CNH属于可交割的远期，即DF；NDF则无法交割本金,实际结算时使用其他可自由兑换的货币）
    + 买报价（bid）&卖报价（ask）：外汇的买/卖报价是针对银行而言的，银行买外汇时使用买价，卖出时使用卖价
