#### 金融产品定价理论
+ 1896 欧文·费雪 & 直接定价法：$PV = \Sigma \frac{CF_{t}}{(1+r)^{t}} $（如何确定贴现率r？）

+ 1952 马科维茨： “证券组合分析” Mean-Var Analysis
+ 1960s 夏普：“资本资产定价模型”（CAPM）$E(r_{i})=r_{f}+β_{im}(E(r_{m})-r_{f})$ $β$衡量系统性风险
+ 1973 布莱克、舒尔斯 & 默顿：B-S期权定价模型
+ 1980s 达菲： 拓展到不完全竞争市场

#### 衍生品定价的基本假设
+ 市场不存在摩擦
+ 市场参与者不承担对手风险（Counterparty Risk） （OTC产品定价变简单）
+ 市场是完全竞争的
+ 市场不存在无风险套利机会，或者说市场参与者认为财富越多越好 （用于检验价格“合理性”）

#### 连续复利（一年计息无数次的利率）
先列出m次计息情况下的终值：
$$A(1+\frac{r_{m}}{m})^{mn}$$
利用高数里面的重要极限：
$$\lim_{x→\infty}(1+\frac{1}{x})^{x} = e$$
整理原式之后，代入极限：
$$(A(1+\frac{1}{\frac{m}{r_{m}}})^{\frac{m}{r_{m}}})^{r_{m}n}  $$
$$A(1+ e^{rn})$$
$A(1+ e^{rn})$表示连续复利情况下的终值

##### 各种计息频次利率与连续利率的转化公式
基础：令终值相等，列等式
$$e^{r} = (1+\frac{r_{m}}{m})^{m} = 1+r_{1}$$
即可推出：
$$r = m ln (1+\frac{r_{m}}{m}) $$

$$r_{m} = m(e^{\frac{r}{m}}-1)$$
尤其是当m = 1时： 
$$r = ln (1+r_{1}) = ln (P_{1})-ln (P_{0})$$

在股票投资场景中，由于对数收益率$ln (\frac{P_{1}}{P_{0}})$比百分比收益率$\frac{P_{1}-P_{0}}{P_{0}}$拥有正态性、可连加性等优良性质，一般对数收益率更加常用

### 远期&期货概述

#### 远期合约种类
1. 远期*利率*协议（FRA）
    + 定义：买卖双方同意从未来某一商定的时刻开始，在某一特定时期内按协议利率借贷一笔数额确定、以特定货币表示的名义本金的协议。
    + 通过事先约定利率，借款人可以规避利率上升的风险，贷款人则可以规避利率下跌的风险。无论未来利率升跌，FRA将实际贷款利率锁定为协议利率。
    + $1*4$：一个月后，借款3个月（至第四个月）
2. 远期*外汇*协议
    + 定义：双方约定在将来某一时间按约定的汇率买卖一定金额的某种外汇的合约
    + CNY:在岸市场的远期外汇交易，标的为在岸即期￥汇率；CNH：离岸市场的远期外汇交易，标的为离岸即期￥汇率；NDF:游离于监管之外的外汇市场，标的为在岸市场的￥汇率（CNY、CNH属于可交割的远期，即DF；NDF则无法交割本金,实际结算时使用其他可自由兑换的货币）
    + 买报价（bid）&卖报价（ask）：外汇的买/卖报价是针对银行而言的，银行买外汇时使用买价，卖出时使用卖价

#### 期货
1. 定义

2. 特点
+ 交易所内集中交易：信息优势，流动性好
    + 未平仓合约数：平仓时减少，建仓时增加
    + 当大部分交易为日内交易时，一天内的期货交易量（Volume of Trading）可能大于该日收盘时的未平仓合约数

+ 特殊的交易和交割制度：控制信用风险
    + 每日盯市结算（Marking to Market and Daily Settlement）
        + 每日盯市结算（mark to market）
        + 收盘价!=结算价
        + $F_{T}→S_{T}$：T时刻正好期货交割，所以期货与现货的价值是等同的，现货价格与期货价格应当相等
        + 基差 = S-F（有时候是F-S）在T时刻收敛于0
            + 投机力量主导时：基差主要受情绪主导
            + 套利力量主导时：基差主要受持仓成本主导

    + 保证金（Margin）制度
        + 初始保证金（initial margin）
        + 每日结算价格（Settlement Price）
        + 维持保证金（maintenance margin）
        + 保证金追加通知（margin call）
        + 当保证金账户余额小于维持保证金时，就需要追加保证金直至余额等于初始保证金
        + 交易保证金率：1）按期货合约价值确定。 一般为合约价值的一定比例。这种方法为我国的期货市场普遍采用。（2）按价格的变动量确定。主要是依据期货合约价格的变动量来确定保证金值。这种方法为国外交易所采用。

##### 标准化的合约条款
+ 交易单位(trade unit)（或合约规模contract size）
    + 合约乘数：用于设置市场进入门槛
+ 到期时间

+ 其他
    + 交易时间、产品代码、头寸限额（position limit，若非有现货者套期保值，一个人不可以持有过多某种合约的头寸）等

##### 与远期的不同（场内衍生品&场外衍生品的区别）

1. 交易场所不同
    + 远期并没有固定的交易场所， 买卖双方各自寻找合适的对象， 因而是一个无严格组织的分散市场。 
    + 期货合约则在交易所内集中交易， 一般不允许场外交易。 交易所不仅为期货交易提供了交易场所和统一清算， 还为期货交易制定了许多严格的交易规则（ 如保证金制度 、 每日盯市结算制度 、涨跌停板制度 、最小价格波动值、 报价方式和最大持仓限额等） ， 并为期货交易提供信用担保。 可以说， 期货市场是一个有组织、有秩序的统一市场。

2. 标准化程度不同
    + 远期交易遵循“ 契约自由” 的原则， 合约中的相关条件如标的资产的质量、 数量、交割地点和交割月份都是根据双方具体的需要确定的， 其条款的具体内容往往体现出很强的个性化特征。 
    + 期货合约则是标准化的。 期货交易所为各种期货合约制定了标准化的数量、质量、交割地点、交割时间、交割方式、合约规模等条款， 只有价格是在成交时根据市场行情确定的。 

3. 违约风险不同
    + 远期合约的履行主要取决于签约双方的信用， 一旦一方无力或不愿履约， 另一方就会蒙受损失。 目前国际主要远期市场的参与者大多需达到一定的资质， 并通过第三方或资产担保等方式降低远期交易的违约风险。
    + 期货合约的履行则由交易所或清算机构提供担保。 交易双方直接面对的都是交易所， 即使一方违约， 另一方也几乎不会受到影响。 

4.  合约双方关系不同
    + 远期合约是交易双方直接签订的， 而且由于远期合约的违约风险主要取决于交易对手的信用， 因此签约前通常要对交易对手的信誉和实力等方面作充分的了解。
    + 期货合约的履行完全不取决于对方， 只取决于交易所或清算机构， 交易所是所有买方的卖者和所有卖方的买者。 在期货交易中， 交易者根本无须知道对方是谁， 市场信息成本很低。

5. 价格确定方式不同
    + 远期合约的交割价格是由交易双方直接谈判并私下确定的。 
    + 期货交易的价格则是在交易所中通过公开竞价或根据做市商报价确定的。

6. 结算方式不同
    + 远期交易是到期一次性结算的，所以在远期存续期内，实际交割价格始终不变，标的资产市场价格的变化给投资者带来的是账面浮动盈亏，到期结算时标的资产的市场价格与交割价格的差异才是投资者的真实盈亏。

    + 期货则有所不同。由于期货是*每日盯市结算*实现真实盈亏的，因此可以把期货看作一个每日以结算价平仓结清并以该结算价重新开立的合约，每日结算价格就是不断变动的期货交割价格。

    + 由于每天盯市结算，所以期货的最后结算价格仍然是T时刻的真实价格$F_{T}$；远期只进行一次结算，按照买入远期时的价格$F_{0}$交割

7. 结清方式不同
    + 由于远期合约是非标准化的， 不易找到转让对象， 并要征得原交易对手的同意， 因此绝大多数远期合约只能通过到期实物交割或现金结算来结束。
    + 期货合约则可以通过到期交割结算和平仓两种方式结清。 由于交易便利， 期货交易一方无须征得对方同意就可通过平仓来结清自己的头寸并把履约权利和义务转让给第三方。在实际中， 绝大多数期货合约都是通过平仓来了结的。

##### 卖空（Short Selling）
+ 出售不拥有的资产
+ 向其他投资者借入该资产并卖出
+ 未来需买回归还
+ 此期间需支付原持有者应获得的股利等收入（红利是要还的）

卖空的限制：
+ A股的“卖空”：融券机制，只有融券名单上的资产可以卖空（借券费用很高！）
+ 提升（uptick rules）卖空需要提高报价

1. 无套利方法主要适用于投资性资产，不一定适用于消费性资产
2. 交割价格（Delivery Price）：合约所规定的固定价格 $K$
3. 远期价值：远期合约本身的价值（作为0成本的零和游戏，公平的远期价值应当是0）
4. 远期价格（Forward Price）：使得远期价值为0的合理交割价格 $F$
+ 当 $K = F$时，远期价值为0， $K$ 是一个比较合理的价格

##### 远期价格&期货价格
1. 当无风险利率恒定且对所有到期日都相同时，其他条件相同的远期价格和期货价格相等。
2. 当利率变化无法预测时，两者略有不同 (P48，涉及保证金账户余额的变化)
    + 当标的资产价格与利率正相关时，期货价格高于远期价格
    + 当标的资产价格与利率负相关时，远期价格高于期货价格
3. 远期价格和期货价格的差异幅度还取决于：
    + 合约有效期、税收、交易费用、保证金的处理方式、违约风险、流动性等

##### 基本假设
1. 没有交易费用和税收。
2. 市场参与者能以相同的无风险利率借入和贷出资金。
3. 允许现货卖空。
4. 没有违约风险。
5. 当套利机会出现时，市场参与者将参与套利活动，从而使套利机会消失。
    + 我们得到的理论价格就是*没有套利机会下的均衡价格*。
6. 期货合约的保证金账户支付同样的无风险利率。
    + 这意味着任何人均可不花成本地取得远期和期货的多头和空头地位
    + 此时期货价格 = 远期价格

+ 前三条假设可以有条件放宽

##### 复制定价法
+ 基本思路：
    + 构建两种投资组合，令其终值相等，则其现值一定相等(MF:无套利引理 2)；否则就可进行套利，即卖出现值较高的投资组合，买入现值较低的投资组合，并持有到期末，套利者就可赚取无风险收益。

    + 众多套利者这样做的结果，将使较高现值的投资组合价格下降，而较低现值的投资组合价格上升，直至套利机会消失，此时两种组合的现值相等。

    + 这样，我们就可根据两种组合现值相等的关系求出远期价格。

+ 构造投资组合：
    + 组合B：一单位标的资产(无红利资产)
        + 终值：$t = T$时，$W_{T} = S_{T}$
        + 现值：$W_{t} = S_{t}$

    + 组合A：一份远期合约多头 & 一笔数额为$K e^{-r(T-t)}$的现金（无风险投资）

        + 该远期合约规定多头在到期日T可按交割价格K购买一单位标的资产。
        + 终值：$t = T$时,$W_{T} = S_{T}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$（f为合约价值，K为交割价）
    
    + 令组合A、B现值相等，可得：$f = S_{t} - K e^{-r(T-t)}$

+ 无红利资产的现货－远期平价定理（spot-forward parity theorem），或称现货－期货平价定理（spot-futures parity theorem）
    + F为使合约价值f为0的交割价K
    + $f=0$时，可得$K= F = S e^{r(T-t)}$

+ 证明（反证法）
    + 若$K>S e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）:
        + t时刻：+ 1单位标的资产S   - 1单位远期合约f

        + T时刻： 交割标的资产（卖出标的资产）：- $S e^{r(T-t)}$    + K

        + $W_{t} = 0$, $W_{T} = K- S e^{r(T-t)} > 0$

    + 若$K<S e^{r(T-t)}$ （资产被低估，套利方向是做多标的资产）:
        + t时刻：+ 1单位远期合约f   - 1单位标的资产S

        + T时刻： 交割标的资产（买入标的资产）：+ $S e^{r(T-t)}$    - K
        
        + $W_{t} = 0$, $W_{T} = S e^{r(T-t)} - K  > 0$

##### 远期价格的期限结构
+ F为T时刻交割的远期；$F^{*}$是$T^{*}$时刻交割的远期
+ 根据定价公式，可得：$F = S e^{r(T-t)}$    $F^{*} = S e^{r^{*}(T^{*}-t)}$    
+ 二者相除消去S，可得：$F^{*} = F e^{r^{*}(T^{*}-t)-r(T-t)}$

##### 标的资产：支付已知红利的资产
+ 支付已知红利资产是指在远期合约到期前会产生完全可预测的现金流的资产，如附息债券和支付已知现金红利的股票等
+ $I = 已知红利的现值$

+ 复制定价：构造投资组合
    + 组合B：
        + 一单位标的资产S & 利率为无风险利率r、期限为从当前时刻到红利派发日(T-t)、本金为I的负债 *（负债项是为了抵消红利）*
        + 现值：$W_{t} = S_{t} - I$

    + 组合A：
        + 一份远期合约多头f & 现金$K e^{-r(T-t)}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$
    
    + 令二者现值相等，解出f：$f = S_{t} - I - K e^{-r(T-t)}$
    
    + 令$f = 0$，解出K：$K = F = (S_{t} - I) e^{r(T-t)}$

+ 证明（反证法）：
    + 若$K > = (S_{t} - I) e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）
        + t时刻：+ 一单位标的资产   - 一份该资产的远期合约

        + t~T时刻：从标的资产获得的现金收益以无风险利率r贷出至T时刻（t时刻现值收益为I，T时刻终值收益为 $I e^{r(T-t)}$）

        + T时刻： 交割标的资产（卖出标的资产）：- $S e^{r(T-t)}$    + K
        
        + $W_{t} = 0$, $W_{T} = K- (S-I) e^{r(T-t)} > 0$



    + 若$K > = (S_{t} - I) e^{r(T-t)}$（资产被低估，套利方向是做多标的资产）
        + t时刻：+ 一份该资产的远期合约   - 一单位标的资产

        + t~T时刻：从标的资产获得的现金收益在T时刻归还（t时刻现值亏损为I，T时刻终值亏损为 $I e^{r(T-t)}$）

        + T时刻： 交割标的资产（买入标的资产）：+ $S e^{r(T-t)}$    - K

        + $W_{t} = 0$, $W_{T} =  (S-I) e^{r(T-t)} - K > 0$

##### 标的资产：支付已知红利率的资产
+ 支付已知红利率资产，是指在远期合约到期前将产生与该资产现货价格成一定比率的收益的资产，如外汇、股票指数等，贵金属需要储藏成本，所以存在*负的红利率*或*负的红利*。

+ $q = 已知红利率$ (以连续复利计算)

+ 复制定价：构造投资组合
    + 组合B：
        + $e^{-q(T-t)}$标的资产S，红利收入用于再投资标的资产   （$e^{-q(T-t)}$是为了抵消红利率，最终到期时为 $S$）
        + 现值：$W_{t} = S_{t} e^{-q(T-t)}$

    + 组合A：
        + 一份远期合约多头f & 现金$K e^{-r(T-t)}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$
    
    + 令二者现值相等，解出f：$f = S_{t} e^{-q(T-t)} - K e^{-r(T-t)}$
    
    + 令$f = 0$，解出K：$K = F = (S_{t}) e^{(r-q)(T-t)}$

+ 证明（反证法）：
    + 若$K >  (S_{t}) e^{(r-q)(T-t)}$（资产被高估，套利方向是做空标的资产）
        + t时刻：+ $e^{-q(T-t)} $单位标的资产   - 1份该资产的远期合约 

        + T时刻： 交割标的资产（卖出标的资产）：- $S e^{(r-q)(T-t)}$    + K

        + $W_{t} = 0$, $W_{T} = K- S e^{(r-q)(T-t)}> 0$

    + 若$K < (S_{t}) e^{(r-q)(T-t)}$（资产被低估，套利方向是做多标的资产）
        + t时刻：+ 1份该资产的远期合约   - $e^{-q(T-t)} $单位标的资产

        + T时刻： 交割标的资产（买入标的资产）：+ $S e^{(r-q)(T-t)}$    - K
        
        + $W_{t} = 0$, $W_{T} =  S e^{(r-q)(T-t)} - K > 0$

##### 持有成本（Cost of Carry）
+ 持有成本 = 保存成本+利息成本- 标的资产在合约期限内的收益
+ 例子：
    + 不支付红利的股票，持有成本是$c=r$
    + 股票指数，持有成本是$c=r - q$
    + 外币，持有成本是$c=r-r_{f}$
    + 黄金白银等，持有成本$c=r+u$

+ 定价公式可以用持有成本表示：
    + $F = S e^{c(T-t)}$
    + $f = S e^{(c-r)(T-t)} - K e^{r(T-t)}$

##### 不完美市场
1. 存在交易成本的时候，假定每一笔交易的费率为Y
    + 远期价格不再是具体值，而是一个区间：$[S(1-Y)e^{r(T-t)},S(1+Y)e^{r(T-t)}]$
    + 相当于在套利判定的不等式中加入了一个“楔子”

2. 借贷存在利差的时候，如果用$r_{b}$表示借入利率，用$r_{l}$表示借出利率，对非银行的机构和个人，一般是$r_{b}>r_{l}$。
    + 此时的价格区间为：$[Se^{r_{l}(T-t)},Se^{r_{b}(T-t)}]$
    + 借入资金的价格上升，卖空投资的收益下降
    
3. 存在卖空限制的时候，假设由于卖空限制导致的成本比例为X。
    + 此时的价格区间为：$[(1-X)Se^{r(T-t)},Se^{r(T-t)}]$
    + 上限不变，不存在限制；卖空存在X的成本

4. 假设以上情况同时存在：$[(1-X)S(1-Y)e^{r_{l}(T-t)},S(1+Y)e^{r_{b}(T-t)}]$

##### 消费性资产的远期定价

+ 远期价格：$F \leq S e^{c(T-t)}$

+ 原因：
    + 消费性资产具有消费价值（通常称之为便利收益），而远期合约却无法即时消费，消费性资产与其远期之间并不具有完全的可替代性。

    + 因此即使在远期相对价值偏低的时候投资者也不会轻易出售现货，购买远期，从而使得单纯基于风险收益考虑的金融无套利原则不再完全有效。
    

##### $F_{t}$和$S_{t}$、$E(S_{t})$的关系



+ 同一时刻远期（期货）价格与标的资产现货价格的关系：
    + 同一时刻的两者价格高低取决于持有成本（$F_{t} = S_{t} e^{c(T-t)}$）；在远期（期货）到期日，远期（期货）价格将收敛于标的资产的现货价格

    + 标的资产的现货价格对同一时刻的远期（期货）价格起着重要的制约关系（案例3.6 “327”国债期货事件）

    + 远期（期货）与现货的相对价格*只与持有成本c有关*，与未来现货的涨跌预期无关。

    + 价格的领先滞后关系（价格发现功能）

+ 预期收益率：$E(S_{t}) = S e^{y(T-t)}$ （当资产价格服从对数正态分布时成立，y表示预期收益率，$y = r + 风险溢酬$）
    + $F_{t}$和$E(S_{t})$的关系由y和r的大小关系决定

![F_E(S)](./f_es.png)

+ 在一个无套利的有效市场中，标的资产和其冗余证券期货之间具有一体化性质，期货的预期收益率总是正好等于标的资产的风险溢酬。

+ 转移风险和管理风险是期货市场的最本质功能。

### 第四章 远期和期货的应用

##### 主要应用：
+ 风险管理
    + 按多/空区分：
        + 多头（买入）风险管理
            + 通过做多远期或期货进行风险管理
            + 适合*担心价格上涨*的投资者，锁定未来买入价格
        + 空头（卖出）风险管理
            + 通过做空远期或期货进行风险管理
            + 适合*担心价格下跌*的投资者，锁定未来卖出价格
    + 按效果区分：
        + 完美的套期保值：
            + 能够完全消除价格风险的套期保值。 （注意：不是指价格不变，而是指未来的价格是确定的）
            + 远期（期货）的到期日、标的资产和交易金额等条件的设定使得远期（期货）与现货恰好匹配。
        + 不完美的套期保值：
            + 无法完全消除价格风险的套期保值。
            + 常态
        + 原因：
            + 基差风险（Basis Risk）
                + 基差：特定时刻被套期保值的现货价格H 与用以进行套期保值的期货价格G之差（此时现货不一定是期货的标的资产，期货价格不一定等于理论价格）
                + b = H − G
                + 套保到期时基差的不确定性导致了不完美的套期保值
            + 数量风险（Quantity Risk）
                + 可能由于事先无法确知需要套期保值的标的资产规模
                + 可能由于期货合约的标准数量无法完全对冲现货的价格风险
                + 讨论*最优套期保值比率*时，通常不考虑数量风险
            + 相比远期，期货更加标准化，更不易实现完美套期保值
            
    + 运用远期或期货进行套期保值，消除了价格风险，但并不保证盈利。
        + 对于企业来说，完全对冲风险并非是最好的
    + 风险管理应该是针对企业的全局风险进行的，而不是针对局部风险。
        + eg. 炼油企业&原油期货：原油上涨，成品油也会上涨，因此企业应该通过做空来降低整体风险

+ 套利
    + 从第三章的定价方法出发，实际价格偏离理论价格即可套利
    + 在实际操作中，要考虑市场中各种“摩擦”

+ 投机
    + 从本质上看，远期（期货）与其标的资产价格变动的风险源是相同的，只是交割时间不同。
        + 因此远期（期货）与其标的资产现货之间往往存在着良好的替代关系，投机者通过承担价格变动的风险获取收益，既可以通过远期（期货）实现，也可以通过现货实现。

        + 远期（期货）优势在于进入成本低，具有高杠杆效应（双刃剑）。
        + 高杠杆：放大收益/放大亏损
        + $杠杆率 = \frac{1}{保证金比率} $
    + 客观上，套期保值者和套利者往往不足以维持一个市场需要的流动性，此时适度投机可以起到提供流动性的作用。
        + 从这个意义上说，远期（期货）投机者不仅通过承担价格变动的市场风险，也通过承担流动性风险来获取风险收益。

##### 不完美套保
+ 完美的套期保值：
    + 能够完全消除价格风险的套期保值。 （注意：不是指价格不变，而是指未来的价格是确定的）
    + 远期（期货）的到期日、标的资产和交易金额等条件的设定使得远期（期货）与现货恰好匹配。
+ 不完美的套期保值：
    + 无法完全消除价格风险的套期保值。
    + 常态
+ 原因：
    + 基差风险（Basis Risk）
        + 基差：特定时刻被套期保值的现货价格H 与用以进行套期保值的期货价格G之差（此时现货不一定是期货的标的资产，期货价格不一定等于理论价格）
            + b = H(现货价格) − G(期货价格)
            + 现货多头&期货空头：$ W_{1} = (H_{1}-H_{0})+(G_{1}-G_{0}) = (H_{1}-G_{1})+(H_{0}-G_{0}) = b_{1}-b_{0} $,$b_{1}$是未知的，$b_{0}$是已知的，是否完美套保取决于$b_{1}$
            + 分解$b_{1}$：$b_{1}= H_{1}-G_{1} = (H_{1}-S_{1})+(S_{1}-G_{1})$,$S_{1}$为标的资产价格，若$H_{1}=S_{1}=G_{1}$,则$b_{1}=0$，达成完美套保
        + 套保到期时基差的不确定性导致了不完美的套期保值
        + 由于$b_{1}$的变动往往小于G的变化，所以不完美套保往往也能对冲风险
        +  $b_{0}-b_{1}$减小→多头套保有利；$b_{1}-b_{0}$增大→空头套保有利
    + 数量风险（Quantity Risk）
        + 可能由于事先无法确知需要套期保值的标的资产规模
        + 可能由于期货合约的标准数量无法完全对冲现货的价格风险
        + 讨论*最优套期保值比率*时，通常不考虑数量风险
        + 相比远期，期货更加标准化，更不易实现完美套期保值

##### 套保策略
+ 合约种类
    + 足够流动性 & 与现货资产高度相关 
+ 到期日
    + 对于实物交割的期货而言，要避免在*期货到期的月份*中持有期货头寸，以防止逼仓。（避免麻烦的交割流程 & 交割月频繁的价格波动）
    + 在到期时间无法完全吻合时，通常选择比所需的套期保值*月份略晚但尽量接近*的期货品种。（避免期货价与现货价差距过大）
    + 所需套期保值时间较长时，可使用套期保值*展期*（快要到期时，平仓原有合约，购买期限更靠后的合约），但可能给套期保值者带来额外的风险。
+ 头寸方向
    + 价格上涨时不利→多头套保
    + 价格下跌时不利→空头套保
+ 交易数量

##### 套保比率
+ 套期保值比率（Hedge Ratio）是指用于套期保值的资产（如远期和期货）头寸对被套期保值的资产头寸的比率
+ $套期保值比率 = \frac{用于套保的资产头寸}{被套保的资产头寸}$
+ 最优套期保值比率：能够最有效、最大程度地消除被保值对象价格变动风险的套期保值比率。（最小的套保比率）
    + 存在基差风险时，最优套期保值比率很可能不为1

##### 最优套保比率（风险最小化的套保比率）

设1单位现货，n单位期货多头套保：

套保组合：$\Delta \Pi = n\Delta G- \Delta H$

最优套保条件(套保组合的价格变化不依赖于现货价格)：$\frac{\partial \Delta \Pi}{\partial \Delta H} = \frac{\partial (n\Delta G- \Delta H)}{\partial \Delta H} = n \frac{\partial \Delta G}{\partial \Delta H} -1 = 0$ → $n = \frac{\partial \Delta H}{\partial \Delta G}$

由于收益率$r_{H} = \frac{H_{1}-H_{0}}{H_{0}} = \frac{\Delta H}{H_{0}},r_{G} = \frac{\Delta G}{G_{0}}$,

所以$n = \frac{\partial \Delta H}{\partial \Delta G} = \frac{H_{0}* \partial r_{H}}{G_{0}* \partial r_{G}}$

上述最优套期保值比率n仅针对单位价值变动，因此，实际的最优套期保值数量N还应在n的基础上考虑具体的头寸规模,此时现货头寸为N

$N = n*\frac{Q_{H}}{Q_{G}} = \frac{\partial (\Delta H)*Q_{H}}{\partial (\Delta G)*Q_{G}} = \frac{Q_{H}*H_{0}* \partial r_{H}}{Q_{G}*G_{0}* \partial r_{G}}$

##### 最小方差套保比率
此时用方差$\sigma_{\Pi}^{2}$表示组合的风险

$\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$

对n求导，令$\frac{\partial (\sigma_{\Pi}^{2})}{\partial n} = 2\sigma_{G}^{2} n -2 \sigma_{HG}^{2} n =0$

由于$\frac{\partial^{2} (\sigma_{\Pi}^{2})}{\partial n^{2}} = 2\sigma_{G}^{2}>0$恒成立，所以最小化$\sigma_{\Pi}^{2}$的n为$n = \frac{\sigma_{HG}^{2}}{\sigma_{G}^{2}}$

##### 最小方差套保比率& OLS

注意到，若对$\Delta H = \beta \Delta G + \alpha + \epsilon$进行OLS线性回归，则有$\beta = \frac{Cov(H,G)}{Var(G)}$

因此，实践中可以通过OLS的$\hat{\beta} = \frac{\Sigma x_{i}y_{i} - N \bar{x}\bar{y}}{\Sigma x_{i}^{2} -N \bar{x}^{2}}$来估计$n$

ΔH 与ΔG 的期间应与实际套期保值期长度相同，且时期之间不宜重合（Overlapping）。但实际应用中ΔH与ΔG多用日度数据，与套保长度不符。

##### 基于收益率的OLS
不用价格，而用收益率进行回归：$r_{H} = b r_{G} + a + \epsilon$

此时套保比率为：$n = \hat{\beta} = \frac{Cov(\Delta H,\Delta G)}{Var(\Delta G)} = \frac{Cov(\Delta H_{0}r_{H},\Delta G_{0}r_{G})}{Var(\Delta G_{0}r_{G})} = \hat{b}\frac{H_{0}}{G_{0}} $

实际套保数量则是：$N = n \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{H_{0}}{G_{0}} \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{V_{H}}{V_{G}} $

##### 套保有效性
最小方差套保比率的有效性可以通过检验风险降低的百分比来确定。因此，定义套期保值有效性为：$e^{*} = \frac{\sigma_{H}^{2}-\sigma_{\Pi}^{2}}{\sigma_{H}^{2}} $

将 $\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$ 代入，化简可得 $e^{*} = \rho_{HG}^{2} = \rho_{r_{H}r_{G}}^{2} = 线性回归(r_{H} = b r_{G} + a + \epsilon) 的判别系数R^{2} $

显然，$e^{*}$越接近1，套保效果越好

+ 实际上，只要是基于同一或是类似标的资产的现货与衍生产品之间，都可以相互进行风险对冲。
    + 例如期货可以用来为相同或类似标的资产的期权空头进行风险对冲。

+ 注：静态套保只适用于线性衍生品。对于债券、期权等非线性（payoff不再只是直线）衍生品，只能用动态套保。在动态套保时不能运用统计方法（线性回归）来求n, 而只能通过理论关系求偏导。


### 第五章 多种期货与远期

##### 股指期货
+  标的资产：股价指数
    +  运用统计学中的指数方法编制而成的、反映股市中总体股价或某类股票价格变动和走势情况的一种相对指标。

+  股指期货
    +  以股价指数作为标的资产的股价指数期货，交易双方约定在将来某一特定时间交收“一定点数的股价指数”的标准化期货合约。

+  特殊性质
    +  现金结算而非实物交割
    +  合约规模非固定
        + $合约规模 = 股指期货的价格点数 × 每个指数点所代表的金额（固定的合约乘数）$

+ 一般定价公式：$F = S e^{(r-q)(T-t)} $ ,q表示红利率

+ 指数套利（Index Arbitrage）
    + “程序交易”（Program Trading）：每次买卖都需要买卖一揽子股票，交易所一般借助程序来进行交易
    + 1987年10月19日“黑色星期一”
        +当时 纽约股票交易所的交易系统超负荷运行，无法处理大量交易指令，指数套利无法进行。 

+ 套期保值
    + 管理系统性风险
    + 多为交叉套期保值


##### 股指期货的最小方差套保比率
+ CAPM：$R_{i} = \beta_{i}(R_{M} - r) + r $，其中 $\beta_{i} = \frac{Cov(R_{i},R_{M})}{Var(R_{M})}$
    →类比：$  \hat{b} = \frac{Cov(\Delta r_{H},\Delta r_{G})}{Var(\Delta r_{G})}$

    + 注意：CAPM中不涉及期货价格，需要用*未来的*$\beta$才能近似$\hat{b}$
+ 假定：
    1. 被套期保值的股票组合与市场指数$R_{M}$ 之间的β系数等于股票组合与股指期货之间的β系数
    2. 所采用的β系数等于套期保值期间真实的β系数
