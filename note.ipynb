#### 连续复利（一年计息无数次的利率）
先列出m次计息情况下的终值：
$$A(1+\frac{r_{m}}{m})^{mn}$$
利用高数里面的重要极限：
$$\lim_{x→\infty}(1+\frac{1}{x})^{x} = e$$
整理原式之后，代入极限：
$$(A(1+\frac{1}{\frac{m}{r_{m}}})^{\frac{m}{r_{m}}})^{r_{m}n}  $$
$$A(1{rn})$$
$A(1{rn})$表示连续复利情况下的终值

##### 各种计息频次利率与连续利率的转化公式
基础：令终值相等，列等式
$$e^{r} = (1+\frac{r_{m}}{m})^{m} = 1+r_{1}$$
即可推出：
$$r = m ln (1+\frac{r_{m}}{m}) $$

$$r_{m} = m(e^{\frac{r}{m}}-1)$$
尤其是当m = 1时： 
$$r = ln (1+r_{1}) = ln (P_{1})-ln (P_{0})$$

在股票投资场景中，由于对数收益率$ln (\frac{P_{1}}{P_{0}})$比百分比收益率$\frac{P_{1}-P_{0}}{P_{0}}$拥有正态性. 可连加性等优良性质，一般对数收益率更加常用

##### 与远期的不同（场内衍生品&场外衍生品的区别）

1. 交易场所不同
    并没有固定的交易场所， 买卖双方各自寻找合适的对象， 因而是一个无严格组织的分散市场。 
    合约则在交易所内集中交易， 一般不允许场外交易。 交易所不仅为期货交易提供了交易场所和统一清算， 还为期货交易制定了许多严格的交易规则（ 如保证金制度 .  每日盯市结算制度 . 涨跌停板制度 . 最小价格波动值.  报价方式和最大持仓限额等） ， 并为期货交易提供信用担保。 可以说， 期货市场是一个有组织. 有秩序的统一市场。

2. 标准化程度不同
    交易遵循“ 契约自由” 的原则， 合约中的相关条件如标的资产的质量.  数量. 交割地点和交割月份都是根据双方具体的需要确定的， 其条款的具体内容往往体现出很强的个性化特征。 
    合约则是标准化的。 期货交易所为各种期货合约制定了标准化的数量. 质量. 交割地点. 交割时间. 交割方式. 合约规模等条款， 只有价格是在成交时根据市场行情确定的。 

3. 违约风险不同
    合约的履行主要取决于签约双方的信用， 一旦一方无力或不愿履约， 另一方就会蒙受损失。 目前国际主要远期市场的参与者大多需达到一定的资质， 并通过第三方或资产担保等方式降低远期交易的违约风险。
    合约的履行则由交易所或清算机构提供担保。 交易双方直接面对的都是交易所， 即使一方违约， 另一方也几乎不会受到影响。 

4.  合约双方关系不同
    合约是交易双方直接签订的， 而且由于远期合约的违约风险主要取决于交易对手的信用， 因此签约前通常要对交易对手的信誉和实力等方面作充分的了解。
    合约的履行完全不取决于对方， 只取决于交易所或清算机构， 交易所是所有买方的卖者和所有卖方的买者。 在期货交易中， 交易者根本无须知道对方是谁， 市场信息成本很低。

5. 价格确定方式不同
    合约的交割价格是由交易双方直接谈判并私下确定的。 
    交易的价格则是在交易所中通过公开竞价或根据做市商报价确定的。

6. 结算方式不同
    交易是到期一次性结算的，所以在远期存续期内，实际交割价格始终不变，标的资产市场价格的变化给投资者带来的是账面浮动盈亏，到期结算时标的资产的市场价格与交割价格的差异才是投资者的真实盈亏。

    则有所不同。由于期货是*每日盯市结算*实现真实盈亏的，因此可以把期货看作一个每日以结算价平仓结清并以该结算价重新开立的合约，每日结算价格就是不断变动的期货交割价格。

    每天盯市结算，所以期货的最后结算价格仍然是T时刻的真实价格$F_{T}$；远期只进行一次结算，按照买入远期时的价格$F_{0}$交割

7. 结清方式不同
    远期合约是非标准化的， 不易找到转让对象， 并要征得原交易对手的同意， 因此绝大多数远期合约只能通过到期实物交割或现金结算来结束。
    合约则可以通过到期交割结算和平仓两种方式结清。 由于交易便利， 期货交易一方无须征得对方同意就可通过平仓来结清自己的头寸并把履约权利和义务转让给第三方。在实际中， 绝大多数期货合约都是通过平仓来了结的。

##### 复制定价法
思路：
    两种投资组合，令其终值相等，则其现值一定相等(MF:无套利引理 2)；否则就可进行套利，即卖出现值较高的投资组合，买入现值较低的投资组合，并持有到期末，套利者就可赚取无风险收益。

    套利者这样做的结果，将使较高现值的投资组合价格下降，而较低现值的投资组合价格上升，直至套利机会消失，此时两种组合的现值相等。

    ，我们就可根据两种组合现值相等的关系求出远期价格。

投资组合：
    B：一单位标的资产(无红利资产)
        ：$t = T$时，$W_{T} = S_{T}$
        ：$W_{t} = S_{t}$

    A：一份远期合约多头 & 一笔数额为$K e^{-r(T-t)}$的现金（无风险投资）

        期合约规定多头在到期日T可按交割价格K购买一单位标的资产。
        ：$t = T$时,$W_{T} = S_{T}$
        ：$W_{t} = f K {-r(T-t)}$（f为合约价值，K为交割价）
    
    合A. B现值相等，可得：$f = S_{t} - K e^{-r(T-t)}$

利资产的现货－远期平价定理（spot-forward parity theorem），或称现货－期货平价定理（spot-futures parity theorem）
    使合约价值f为0的交割价K
    =0$时，可得$K= F = S e^{r(T-t)}$

（反证法）
    K>S e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）:

        \begin{array}{c|c|c|c|c}
        t & S \rightarrow -S & -S * B \rightarrow S & -1 * f \rightarrow 0 & 0 \\
        \hline
        T & - 1*S \rightarrow 0 & +S * B \rightarrow -S e^{r(T-t)} & +1 * f \rightarrow K & K- S e^{r(T-t)} > 0
        \end{array}
        刻：套利者按无风险利率r借入现金S，期限为T-t。然后用S购买一单位标的资产，同时卖出一份该资产的远期合约，交割价格为K。
        刻：该套利者将一单位标的资产交割换得现金K，并归还借款本息 $S*e^{r(T-t)}$，实现 $K - S*e^{r(T-t)}$ 的无风险利润

    K<S*e^{r(T-t)}$ （资产被低估，套利方向是做多标的资产）:

        \begin{array}{c|c|c|c|c}
        t & - 1*S \rightarrow S & +S * B \rightarrow -S & +1 * f \rightarrow 0 & 0 \\
        \hline
        T & S \rightarrow 0 & -S * B \rightarrow S e^{r(T-t)} & -1 * f \rightarrow -K &  S e^{r(T-t)}-K > 0
        \end{array}
        刻：套利者卖空标的资产获得现金S，以无风险利率r进行投资，期限为T-t。同时买进一份该标的资产的远期合约，交割价为K。
        刻：套利者收到投资本息 $S*e^{r(T-t)}$，并以现金K交割换得一单位标的资产，用于归还卖空时借入的标的资产，从而实现 $S*e^{r(T-t)}-K$ 的无风险利润。
    列为时间点，第二至第四列为套利策略，第五列为总和现金流

##### 远期价格的期限结构
T时刻交割的远期；$F^{*}$是$T^{*}$时刻交割的远期
定价公式，可得：$F = S e^{r(T-t)}$    $F^{*} = S e^{r^{*}(T^{*}-t)}$    
相除消去S，可得：$F^{*} = F e^{r^{*}(T^{*}-t)-r(T-t)}$

    B：
        位标的资产S & 利率为无风险利率r. 期限为从当前时刻到红利派发日(T-t). 本金为I的负债 *（负债项是为了抵消红利）*

    A：
        远期合约多头f & 现金$K e^{-r(T-t)}$
        ：$W_{t} = f K {-r(T-t)}$
    
    者现值相等，解出f：$f = S_{t} - I - K e^{-r(T-t)}$
    
    f = 0$，解出K：$K = F = (S_{t} - I) e^{r(T-t)}$

（反证法）：
    $K > (S_{t} - I) e^{r(T-t)}$（资产被低估，套利方向是做多标的资产）

        \begin{array}{c|c|c|c|c}
        t & S \rightarrow -S & -S * B \rightarrow S & -1 * f \rightarrow 0 & 0 \\
        \hline
        T & - 1*S \rightarrow I * e^{r(T-t)} & -S * B \rightarrow -S e^{r(T-t)} & 1 * f \rightarrow K &  K-(S-I) e^{r(T-t)} > 0
        \end{array}
        刻：套利者按无风险利率r借入现金S，期限为T-t。然后用S购买一单位标的资产，并卖出一份该资产的远期合约，交割价格为K。同时，将T-t期间从标的资产获得的现金收益以无风险利率r贷出至T时刻。
        刻：该套利者将一单位标的资产交割换得现金K，并归还借款本息 $S*e^{r(T-t)}$，同时得到 $I*e^{r(T-t)}$ 的本利收入，最终实现 $K- (S-I)*e^{r(T-t)}$ 的无风险利润

    $K < (S_{t} - I) e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）

##### 套保比率
保值比率（Hedge Ratio）是指用于套期保值的资产（如远期和期货）头寸对被套期保值的资产头寸的比率
期保值比率 = \frac{用于套保的资产头寸}{被套保的资产头寸}$
套期保值比率：能够最有效. 最大程度地消除被保值对象价格变动风险的套期保值比率。（最小的套保比率）
    基差风险时，最优套期保值比率很可能不为1

##### 最优套保比率（风险最小化的套保比率）

设1单位现货，n单位期货多头套保：

套保组合：$\Delta \Pi = n\Delta G- \Delta H$

最优套保条件(套保组合的价格变化不依赖于现货价格)：$\frac{\partial \Delta \Pi}{\partial \Delta H} = \frac{\partial (n\Delta G- \Delta H)}{\partial \Delta H} = n \frac{\partial \Delta G}{\partial \Delta H} -1 = 0$ → $n = \frac{\partial \Delta H}{\partial \Delta G}$

由于收益率$r_{H} = \frac{H_{1}-H_{0}}{H_{0}} = \frac{\Delta H}{H_{0}},r_{G} = \frac{\Delta G}{G_{0}}$,

所以$n = \frac{\partial \Delta H}{\partial \Delta G} = \frac{H_{0}* \partial r_{H}}{G_{0}* \partial r_{G}}$

上述最优套期保值比率n仅针对单位价值变动，因此，实际的最优套期保值数量N还应在n的基础上考虑具体的头寸规模,此时现货头寸为N

$N = n*\frac{Q_{H}}{Q_{G}} = \frac{\partial (\Delta H)*Q_{H}}{\partial (\Delta G)*Q_{G}} = \frac{Q_{H}*H_{0}* \partial r_{H}}{Q_{G}*G_{0}* \partial r_{G}}$

##### 最小方差套保比率
此时用方差$\sigma_{\Pi}^{2}$表示组合的风险

$\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$

对n求导，令$\frac{\partial (\sigma_{\Pi}^{2})}{\partial n} = 2\sigma_{G}^{2} n -2 \sigma_{HG}^{2} n =0$

由于$\frac{\partial^{2} (\sigma_{\Pi}^{2})}{\partial n^{2}} = 2\sigma_{G}^{2}>0$恒成立，所以最小化$\sigma_{\Pi}^{2}$的n为$n = \frac{\sigma_{HG}^{2}}{\sigma_{G}^{2}}$

##### 最小方差套保比率& OLS

注意到，若对$\Delta H = \beta \Delta G + \alpha + \epsilon$进行OLS线性回归，则有$\beta = \frac{Cov(H,G)}{Var(G)}$

因此，实践中可以通过OLS的$\hat{\beta} = \frac{\Sigma x_{i}y_{i} - N \bar{x}\bar{y}}{\Sigma x_{i}^{2} -N \bar{x}^{2}}$来估计$n$

ΔH 与ΔG 的期间应与实际套期保值期长度相同，且时期之间不宜重合（Overlapping）。但实际应用中ΔH与ΔG多用日度数据，与套保长度不符。

##### 基于收益率的OLS
不用价格，而用收益率进行回归：$r_{H} = b r_{G} + a + \epsilon$

此时套保比率为：$n = \hat{\beta} = \frac{Cov(\Delta H,\Delta G)}{Var(\Delta G)} = \frac{Cov(\Delta H_{0}r_{H},\Delta G_{0}r_{G})}{Var(\Delta G_{0}r_{G})} = \hat{b}\frac{H_{0}}{G_{0}} $

实际套保数量则是：$N = n \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{H_{0}}{G_{0}} \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{V_{H}}{V_{G}} $

##### 套保有效性
最小方差套保比率的有效性可以通过检验风险降低的百分比来确定。因此，定义套期保值有效性为：$e^{*} = \frac{\sigma_{H}^{2}-\sigma_{\Pi}^{2}}{\sigma_{H}^{2}} $

将 $\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$ 代入，化简可得 $e^{*} = \rho_{HG}^{2} = \rho_{r_{H}r_{G}}^{2} = 线性回归(r_{H} = b r_{G} + a + \epsilon) 的判别系数R^{2} $

显然，$e^{*}$越接近1，套保效果越好

+ 实际上，只要是基于同一或是类似标的资产的现货与衍生产品之间，都可以相互进行风险对冲。
    + 例如期货可以用来为相同或类似标的资产的期权空头进行风险对冲。

+ 注：静态套保只适用于线性衍生品。对于债券. 期权等非线性（payoff不再只是直线）衍生品，只能用动态套保。在动态套保时不能运用统计方法（线性回归）来求n, 而只能通过理论关系求偏导。

+ 可以利用股指期货，根据自身的预期和特定的需求改变股票投资组合的beta系数（由β 变为β*），从而调整股票组合的系统性风险与预期收益
+ 此时套保比率为 $n = (\beta^{*} - \beta)$
+ 套保数量为 $N = (\beta^{*} - \beta) \frac{V_{H}}{V_{G}}$

#### 外汇远期

+ 外汇远期的持有成本为$r_{f}$,即持有外汇时获得的外国银行无风险利率
##### 远期价值&远期汇率

1. 远期价值：$f = S e^{(r_{f})(T-t)} - K e^{r(T-t)} $
2. 远期汇率（价格）：$F = S e^{(r-r_{f})(T-t)} $
+ 利率平价关系：
    + 若 $r_{f} > r$ , 外汇远期贴水 $F_{t}<S_{t} $；
    + 若 $r_{f} < r$ , 外汇远期升水 $F_{t}>S_{t} $。
    + 注：远期升贴水是指远期汇率与当前即期汇率的相对高低，并不意味着外汇真实的升值与贬值

#### 远期利率
+ 远期利率是指现在时刻 $(t+t^{*}时刻)$的将来一定期限 $(T+t^{* }时刻)$ 的利率。 
    + $([t+t^{*},T+t^{*}])$，$(t+t^{*})*(T+t^{*})$表示 $t+t^{*} $时刻起 $(T-t)$期间的利率
+ 即期利率是指当前时刻(t时刻)起一定期限(T时刻)的利率。 $([t,T])$

设利率期限结构如下：

![](./interest.png)

为了使$T^{*}$时刻终值相等，可得：$$e^{[r(T-t)+r_{F}(T^{*}-T)]} = e^{r^{*}(T^{*}-t)}$$
因此：$$r(T-t)+r_{F}(T^{*}-T) = r^{*}(T^{*}-t)$$（对比：远期价格的期限结构）
$$r_{F} = \frac{[r^{*}(T^{*}-t) - r(T-t)]}{(T^{*}-T)}$$
化简可得：$$r_{F} = (r^{*}-r)\frac{T-t}{T^{*}-T} +r^{*} $$
+ 如果即期利率期限结构在 $T^{*}-T$期间是向上倾斜的，即$r^{*}>r$，则 $r_{F}>r^{*}$（升水）； 反之，则$r_{F}<r^{*}$（贴水）。


##### 远期利率协议（FRA）
+ 远期利率协议（FRA）是买卖双方同意从未来某一商定的时刻开始的一定时期内按*协议利率*借贷一笔数额确定. 以具体货币表示的*名义本金*的协议。
    + 在t时刻与银行约定，在T时刻向银行以$r_{F}$利率借款/存款，在$T^{*}$时刻还款/提款

+ FRA的特征
    +  在T时刻进行现金结算，
        + 结算金额为 $T^{*}$时刻利差贴现至 $T$时刻的贴现值
        + 贴现率为真实利率 $r$
        + 由亏损方支付给盈利方
        + 现金结算：只交割差价，不实际交割贷款

    +  名义本金：合约中的本金只是名义本金，并不一定涉及贷款的实际交付

    +  头寸：Long/Short
        + Long: Fixed-rate payer (固定利率支付者即为FRA的多头)

    +  报价：3×9（3月后，借6个月） SHIBOR（约定的市场参考利率） 2.86（合同的约定利率）
    
+ 注：
    + FRA的买方是名义借款人，其订立FRA的目的主要是为了规避利率上升的风险。
    + FRA的卖方是名义贷款人，其订立FRA的目的主要是为了规避利率下降的风险。

##### FRA定价
1. t时刻存在2个期限相同$(T,T^{*})$的FRA，一个FRA的协议利率为市场协议利率$r_{F}$，一个为$r_{K}$
2. 由于第一个FRA协议利率为$r_{F}$，故$f=0$
3. 第二个FRA的价值即为二者*支付利息*差额的贴现值：$f = e^{-r^{*}(T^{*}-t)}(A e^{r_{F}(T^{*}-T) - r_{K}(T^{*}-T)}) $

##### 期货利率&远期利率
+  欧洲美元期货合约与远期利率协议都锁定了未来一定期限的利率。
+  1 年以下的到期期限，期货利率≈ 远期利率

+  长期：差异不能忽略
    +  一次性到期/每日盯市结算和保证金：远期利率较低（每日盯市可以不采用自融资策略，更加灵活）
    +  盈亏结算时贴现/无贴现：远期利率较低


1. 远期利率协议报出的是远期利率，而利率期货报出的通常并非期货利率，而是与期货利率反向变动的特定价格，期货利率隐含在报价中。
2. 由于上述区别，利率期货结算金额为T时刻协议价与结算价之差，远期利率的结算金额则为T*时刻利差的贴现值。
3. 利率期货存在每日盯市结算与保证金要求，加上结算金额计算方式的不同，决定了远期利率与期货利率的差异。
4. 在远期利率协议中的多头是规避利率上升风险的一方，而利率期货的多头则是规避期货价格上升风险，即规避利率下跌风险的一方。
5. 远期利率协议通常采用现金结算，而利率期货可能需要实物交割，期货交易所通常规定多种符合标准的不同证券均可用以交割，使得利率期货相对复杂

##### 国债现货的全价&净价
+ 报价时通常报出面值每100元的价格
+ $全价=净价+应计利息$,国债现货使用净价而非全价来报价
    + 应计利息：上一个付息日以来的利息（按比例计算），是债券的前持有人应得的利息       
    + $ 100元面值现货的应计利息 = \frac{票面利率*100}{每年付息次数} * \frac{现货交割日(t)-上一个付息日}{当前付息周期实际天数} $

+ 付息周期为 $[上一个付息日,下一个付息日)$，算头不算尾

##### 可交割券与标准券
+  可交割券：期货空方可以选择发行期限不高于7年，合约*到期月份首日*剩余期限为4-5.25年的任何记账式附息国债均为该期货合约的可交割券。

    +  一种现券可以是多种期货的可交割券，一个期货也可以有多种可交割券
    
+  标准券：面值为100万，息票率为 3% ，在交割月首日的剩余到期期限为5年整的虚拟债券（作为比较标准，并不存在于市场交易），是其他实际可交割债券价值的衡量标准。
    +  主要特征：虚拟国债. 整数期限. 票面利率3%. 面值100万
    +  （以3%贴现时，为平价债券），在交割月首日，不存在应计利息，标准券的净价=全价=100


##### 转换因子
+  转换因子：$转换因子 = \frac{可交割券j的净价}{标准券净价} = \frac{可交割券j的净价}{100} $
    + （因为标准券为平价债券且不含应计利息）
+  中金所规定转换因子计算：每1元面值的可交割债券的未来现金流，按3%的年到期收益率贴现到交割月首日的价值，再扣掉该债券每1元面值应计利息后的余额。
    +  时间调整（只算月份之差）
    +  本质上是净价（贴现现金流-应计利息）
    +  交易所公布
+  转换因子的具体计算不作要求（课外阅读）
    
+ Note: 
    1. 同一个债券对于不同国债期货合约的转换因子是不同的；
    2. 给定国债期货的可交割券的转换因子不变。


##### 国债期货的净价和全价
1. 国债期货使用标准券期货净价报价，因此市场上看到的国债期货行情价格就是*标准券*的期货净价

2. $可交割券的期货净价_{j} = 标准券期货净价*转换因子_{j}$

3. $国债期货空方交割时收到的现金 = 期货交割全价 = 可交割券的期货全价_{j} = 可交割券的期货净价_{j} + 期货应计利息_{j} = 标准券期货净价*转换因子_{j} + \frac{可交割券票面利率*100}{每年付息次数}*\frac{配对缴款日(T+2)-上一个付息日}{付息周期实际天数}$

    + 其中$ 期货应计利息 = \frac{可交割券票面利率*100}{每年付息次数} * \frac{配对缴款日(T+2)-上一个付息日}{付息周期实际天数} $
    + 此时的T(交割意向申报日/交割日)只能是交易日（周末不交易）
    
    + T+1：卖方交券日，T+2：配对缴款日，T+3：买方收券日

##### 转换因子的不足
+ 理想的转换因子应该是各债券在交割时刻的真实价格之比
+ 现实中转换因子的不足：
    1. 由于事先无法预知可交割券在交割月的真实到期收益率，计算转换因子时所有债券使用了同一个贴现率3%，这将导致转换因子出现误差；
    2. 在计算转换因子时，由于无法预知真实交割时刻，中金所规定使用整数月份，这进一步使得转换因子成为一个近似值；
    3. 在计算转换因子时，中金所对一年支付一次利息和一年支付两次利息的债券都使用3%的贴现率，实际上应对计息频率进行转换才能使两个贴现率等价。
+ 由于转换因子诸多假设，天然导致不同债券之间并非完美公平转换——相对合算和不合算
+  Question:作为有选择权的卖方，会选哪个券交割？

##### 最合算交割券（CTD券，cheapiest to deliver）
+ 最合算交割券：购买交割券所付的价格与交割期货时空方收到的现金之差最小的债券。
+ 目标：交割日的交割成本最小
    + $ 交割成本_{j} = 现券全价_{j} - 期货全价_{j} = 现券报价_{j} -期货结算价*交割券转换因子_{j} $
    + （T时刻，现券和期货的应计利息基本相等）
+ 准CTD券
    + 因为只有到期货交割时我们才能获知交割时的现券报价和期货结算价。这意味着在国债期货到期交割前，哪个券是真正的CTD券是无法预知的。
    + 所以只能退而求其次，根据当时的信息条件判断哪个券未来交割时最可能成为CTD券，即在国债期货到期交割前，我们只能确定“准CTD券”。市场情况变化，准CTD券也可能发生变化。
    + 判断方法：IRR方法 & 经验方法

##### 国债期货定价
+ 假设：
    1. 不存在择券期权
    2. 不存在择时期权
        + 因为这两个权利属于空方，所以理论定价会高于实际价格
    3. 准CTD券和交割日期已知

+ 定价公式：
    + 由于准CTD券和交割日期已知，所以可得 $S_{t} = 现货净价 + 应计利息 $
    + 所以可得期货全价： $F_{t} = (S_{t}-I_{t}) e^{r(T-t)} $
        + $I = t~T时间段的付息贴现至t时刻$
    + 已知：$期货全价^{j} = 期货净价 * 转换因子^{j}+ 期货应计利息^{j} $
        + 所以 $期货净价 = \frac{期货全价^{j}- 期货应计利息^{j}}{转换因子^{j}} = \frac{F_{t}^{j}- 期货应计利息^{j}}{转换因子^{j}}$
        + 期货应计利息：前一个付息日至T+2时刻的应计利息

##### 计算国债期货的报价
+ 已知：准CTD券的现货报价（净价）以及转换因子. 预期的交割日期
+ t:定价日, T:交割日, T+2:配对缴款日

2. 预期的交割日期 → 合约期限内的现金红利 → $ F = S e^{r(T-t)} $ → 准CTD券的期货全价
    + $ I = \Sigma \frac{CF_{i}}{(1+r)^{i}} $
    + 期限为$[t,T)$

3. 预期的交割日期 → 期货应计利息 → 准CTD券的期货净价
    + $ 期货应计利息 = \frac{100*票面利率}{付息次数}*\frac{配对缴款日-上一个付息日}{下一个付息日-上一个付息日} $

4. 准CTD券的转换因子 → 标准券期货净价（国债期货报价）


##### 国债期货定价存在的问题
1. 由于国债期货卖方拥有择券期权和择时期权，本章所定出的国债期货价格通常会高于真实的国债期货价格。

2. 国债期货只能施行正向套利策略：
    + 对于国债期货来说，只有在国债期货市场价格过高时，实施“买现货卖期货”的套利策略才可能盈利；
    + 但在国债期货市场价格过低时，投资者并不能像普通期货那样进行“买期货卖现货”的套利交易，因为*未来期货交割时间和交割券种的选择权*都在期货空方手中，套利者无法确定赚取无风险利润。

3. 国债期货价格经常小于债券现货价格
    + 期货存续期内现货可能支付票息
    + 现券票息经常高于短期利率
    + 国债期货空方有隐含期权
    + $国债期货价值 = 期货合约价值-择券期权价值-择时期权价值$

4. 对于每一个可交割券，其对应的国债期货基差定义为：
    + 用净价定义：
    + $ 国债期货基差_{j} = 可交割券净价_{j} - 国债期货报价 * 转换因子_{j}$

##### 资产价值的利率风险
+ 资产的利率风险一般被表述为资产价格变动的百分比对到期收益率变动的敏感性。
+ 由于资产价值P是到期收益率y的函数, $P = f(y)$，资产价值的利率风险可以用泰勒展开表示如下：
    + $ \frac{d P}{P} = \frac{1}{P} * \frac{dP}{dy} + \frac{1}{P*2!} * \frac{d^{2}P}{d y^{2}} +...$
    + 其中第一项为久期(duration)，第二期为凸性

##### 久期
+ 久期(Duration)：资产价格变动的百分比对到期收益率变动的一阶敏感性
    + $D = - \frac{d P/P}{dy} = -\frac{1}{P} \frac{d P}{dy} $

+ 久期*一般*为正，此时利率上升，债券价格下降

+ 久期反映了资产价格利率风险的*主要*部分，并非全部，还存在二阶项和高阶项

+ 久期越大，资产的利率风险越大；反之则越小
    + 久期越大，债券价格对利率变动越敏感，利率风险越大

+ 如果资产价值是利率的*非线性函数*，显然其久期并非常数，而是会*动态变化*的

##### 货币久期（美元久期）
+ 货币久期(Dollar Duration)：到期收益率变动引起的价值变动金额：
    + $\$D = D * P= \frac{dP}{dy} $
+  对于非线性产品而言，货币久期也是*时变*的。
+  对衍生品而言，货币久期尤为重要。
    + 因为很多衍生品刚签约时的*合约价值为零*，无法计算合约的久期，只能计算合约的货币久期。
+  对于期货合约来说，由于期货合约价值的变动源于期货价格的变动，因此如果不考虑合约规模的影响，期货合约的货币久期就等于期货价格的货币久期。

##### 麦考利久期
+ 假定债券价格为现金流贴现值： 
    + $P = \Sigma_{i=1}^{n} \frac{CF_{i}}{(1+y)^{i}} +\frac{A}{(1+y)^{n}} $
+ 计算久期D: $D = -\frac{1}{P} \frac{dP}{dy} = \frac{1}{1+y} *[\Sigma \frac{CF_{i}}{(1+y)^{i}} + \frac{A}{(1+y)^{n}}]*\frac{1}{P}  $
    + 其中：$[\Sigma \frac{CF_{i}}{(1+y)^{i}} + \frac{A}{(1+y)^{n}}]*\frac{1}{P} $ 即为麦考利久期
        + 意义：按现金流加权的平均到期期限
    + 真正衡量风险的久期为 $ D$，被称为修正久期

+ 常见误解：
    + ×：麦考利久期就是不含权债券的久期 
        + √：修正久期才是不含权债券的久期
    + ×：久期一定是期限的加权 
        + √：久期与期限加权之间并没有必然联系
    + ×：久期的单位一定是年 
        + √：久期的单位并不必然为年

+ $ D = -\frac{1}{P} \frac{dP}{dy}$才是最具一般性. 最能反映久期性质的定义与公式
+ 连续复利情况下的定价：
    + $P_{t} = \Sigma{CF_{i}}{e^{-y(t_{i}-t)}}$
    + 对其求导，可得久期 $ D_{t} = -\frac{1}{P} \frac{dP}{dy} = -\frac{\Sigma{CF_{i}}{e^{-y(t_{i}-t)}}}{P}  (t_{i}-t)$
    + 此时麦考利久期等于修正久期
    + 在使用连续复利到期收益率时，普通债券的久期就是现金流的加权平均期限，无需再调整

##### 有效久期
+ 用于不存在解析解的定价模式
+ $D = \frac{P_{-}-P_{+}}{2*P*\Delta y} $
    + $P_{-} $与$P_{+} $分别代表到期收益率下跌和上升时所达到的资产价格
    + 核心方法：中心差分

##### 基点价格值
+ 基点价格值(DV01或BPV）
    +  利率变化1个基点引起的资产价格变动
    +  与1个基点的货币久期十分接近
+ $ BPV = \frac{P_{-0.01\%}-P_{+0.01\%}}{2} $
    + 用差分取代微分

+ 利率远期和利率期货的利率敏感性估计取决于其标的资产的利率敏感性和远期（期货）本身价值变化的计算方式。
    + 例如：欧洲美元期货通常使用基点价格值这一利率敏感性指标，即1个基点对应着一份欧洲美元期货合约的价格变动为25美元。

##### 国债期货的久期

+ 回忆：国债期货的定价
    + $G = \frac{期货全价_{准CTD} - 期货应计利息_{准CTD}}{转换因子_{准CTD}}$
    + $期货全价_{准CTD}F = (S-I)e^{r(T-t)} $

1. 准CTD券期货价格F的久期和货币久期
    + 假设：到期收益率y通过影响现货价格S来影响F，y与利率r无关
    1. 已知$ F= (S-I) e^{r(T-t)} $
    2. 货币久期:$$ \$D_{F} = -\frac{dF}{dy} = -e^{r(T-t)}\frac{dS}{dy} = e^{r(T-t)}\$D_{S} $$
    3. 久期:$$ D_{F} = -\frac{1}{F} \frac{dF}{dy} = -\frac{e^{r(T-t)}}{(S-I) e^{r(T-t)}} \frac{dS}{dy} = -\frac{1}{(S-I)} \frac{dS}{dy} \approx -\frac{1}{S} \frac{dS}{dy} = D_{S} $$

2. 国债期货报价G的久期和货币久期
    + 久期：
    $$ D_{G} = -\frac{1}{G}\frac{\delta G}{\delta y} = -\frac{1}{\frac{F -应计利息}{转换因子}}*\frac{\delta ({\frac{F -应计利息}{转换因子}})}{\delta y} = -\frac{1}{F -应计利息} * \frac{\delta F}{\delta y} $$
    $$D_{G} = -\frac{1}{F -应计利息} * D_{F}*F \approx -\frac{1}{F} * D_{F}*F = D_{F} = D_{S} $$
    + 此时G的久期≈准CTD券的久期
    + 货币久期：
    $$ \$D_{G} = -\frac{\delta G}{\delta y} =\frac{\delta ({\frac{F -应计利息}{转换因子}})}{\delta y} = -\frac{1}{转换因子}\frac{\delta F}{\delta y} = \frac{1}{转换因子}\$D_{F} \approx \frac{1}{转换因子}\$D_{S} $$
    + 在现实中运用国债期货进行利率风险管理时，我们是在用期货合约价值的变动去对冲其他资产的利率风险，而每天期货合约价值的变动是由标准券期货报价的变动引起的。因此需要计算国债期货报价的久期和货币久期
+ 注意：
    1. 由于债券和国债期货价格都是利率的非线性函数，其久期和货币久期都是动态变化的；
    2. 国债期货报价的久期和货币久期都是基于准CTD券的利率敏感性指标计算得到的。准CTD券一旦变化，期货的利率敏感性也会随之变化。

##### 基于久期的利率风险管理

+ 由于利率敏感性资产价格与利率存在非线性关系，因此无法进行静态套保，只能进行动态套保。
+ 最优的利率风险套期保值比率n是使得*整个套保组合价值*（而不是其变动率）对利率的一阶敏感性为零的套期保值比率: $\frac{d\Pi}{dy}=0$
    + (目标： $arg_{min}\frac{d\Pi}{dy}$)
+ n实际上是使得套期保值组合的*货币久期*或基点价格值（BPV）为零的套期保值比率

以现货多头和期货空头的空头套期保值组合为例:
+ $d\Pi = dH-n*dG $

+ 令 $\frac{d\Pi}{dy} =0$,可得 $n = \frac{\frac{dH}{dy}}{\frac{dG}{dy}} = \frac{D_{H}*H}{D_{G}*G}$
+ 可得最优套保数量N：$N = \frac{Q_{H}}{Q_{G}}*n = \frac{D_{H}*V_{H}}{D_{G}*V_{G}}$


有时候，并不需要将久期 $D_{H}$对冲至0，只需要将其控制于一个基准 $D_{H}^{*}$

+ 设投资组合的原久期为 $D_{H}$，目标久期为 $D_{H}^{*}$，则需要交易的利率敏感性证券的份数N为
    + $ N = \frac{D_{H}^{*}-D_{H}}{D_{G}}*\frac{V_{H}}{V_{G}}$
    + $V_{G}$ 是一份期货合约按标准券报价计算的合约规模
    + 上式为负时，说明需要进行反向操作，即原来是多头的要卖出，原来是空头的要买进


##### 久期的局限性

+ 久期有着天然的局限性：
    + 久期仅仅是资产价格对利率的一阶敏感性，无法反映和管理资产价格的全部利率风险，当利率变化较大时这个缺陷尤其显著；
        + 有时候需要考虑凸性
    + 久期的定义建立在利率曲线发生平移，即所有期限的利率变化幅度相等的假设基础之上，这是一个不符合现实的假设
        + 没有考虑不同期限的收益率曲线形状不同的情况

### 第六章 互换

+ 互换（Swaps）是两个或两个以上当事人按照商定条件，在约定的时间内交换一系列现金流的合约。
+ 互换可以看作是一系列远期的组合。
+ 最重要和最常见的互换：
    + 利率互换（ Interest Rate Swap，IRS）
    + 货币互换（ Currency Swap）
    + 总收益互换（Total Return Swap，TRS）
    + 信用违约互换（Credit Default Swap，CDS）

##### 利率互换

+ 利率互换，是指双方同意在未来的一定期限内根据同种货币的*相同名义本金*交换现金流的一种*场外交易*的金融合约，其中一方的现金流根据事先选定的某一*浮动利率*计算，而另一方的现金流则根据*固定利率*计算。
+ 常见期限包括1年. 2年. 3年. 4年. 5年. 7年与10年，30年与50年的互换也时有发生。
+ 常见的互换频率包括季度. 半年和年度

案例：国内第一笔利率互换

+ 协议时间：2007年1月18日
+ 交易双方：兴业银行与花期银行
+ 利率互换期限：自2007.1.18起期限为1年
+ 名义本金：未披露（假设为1元）
+ 结算方式：现金结算，本金不需要交换（降低风险敞口）
+ 约定：每三个月交换一次现金流如下：
    + $ 兴业银行→^{\frac{2.808\%}{4}}花旗银行$
    + $ 兴业银行^{\frac{3月期SHIBOR}{4}}←花旗银行$

#### 货币互换
+ 定义
    + 货币互换，是在未来约定期限内将一种货币的本金和固定利息与另一货币的等价本金和固定利息进行交换
    + $货币互换(currency \space swaps) \neq 外汇掉期(foreign \space exchange \space swaps)$
        + 外汇掉期锁定了相对汇率
+ 利率互换与货币互换之差异：
    + 利率互换：通常无需交换本金，只需定期交换利息差额；
    + 货币互换：期初和期末须按照约定的汇率交换不同货币的*本金*（因为名义本金的币种不同），期间还需定期交换不同货币的利息。

+ 流程：
    + ![货币互换](currency_swap.jpg)
    + 假设汇率不变，此时雷公司的实际借款成本为$欧元利率+（支付的美元利率-收到的美元利率） = 4.35%+(6.5%-6.1%) = 4.75% $
    + 由于其中有一部分美元利息，所以雷公司仍要承担汇率风险

#### 总收益互换 TRS

+ 总收益互换：Total Return Swap
    + 在未来约定期限内将一种或一篮子资产的总收益（包括现金流与资本利得或损失）与等值浮动利率债券的利息加或减价差进行交换。
    + $ 支付方 {}^{资产总收益率}\rightleftarrows_{浮动利率+一定的净息差} 接收方 $

+ 特点：
    + 总收益互换的一个重要特征是合约双方并不转让资产的实际所有权，只是转让*收益权*。
        + 对于支付方来说，相当于已经把资产卖掉了，但是不会体现在资产负债表上
        
    + 总收益互换由总收益端（称为总收益支付方，或者总收益互换*卖方*）和融资端（称为总收益接收方，或者总收益互换*买方*）组成。
    + 总收益端的资产可以是单一股票. 指数. 公司债. 抵押贷款. 不动产. 大宗商品. 外汇. 贵金属等。
        + 范围很广

    + 总收益互换优点：

        + 节约交易成本

            + 可以规避买卖资产带来的价格风险和交易费用

        + 提高杠杆

            + 可以实现融资融券的效果

                + 融资（借钱买股）：一边当买方，一边当卖方

                + 融券（借股卖钱）：用TRS卖出手上的股票

        + 规避监管

            + 方便在全球配置资产
            
        + 注意：需用保证金控制风险

#### 信用违约互换 CDS
+ 信用违约互换：Credit Default Swap
    + CDS的买方以一定的费用为代价将基础资产的信用风险转移给CDS的卖方，相当于购入了一份信用保险
    + $ 基础资产 \rightarrow CDS买方 \space \space {}^{支付保费}\rightleftarrows_{发生信用事件时进行赔付} CDS卖方 $
    + 相当于担保信用的保单
+ The Big Short

##### 回购交易
+ $t=0$：$ 正回购方 {}^{债券}\rightleftarrows_{资金Y} 逆回购方 $
+ $t=T$：$ 正回购方 {}^{债券}\rightleftarrows_{资金Y'} 逆回购方 $
+ 显然：$Y'>Y$,相当于支付了回购利息
+ 分类：
    + 质押式回购：债券所有权不发生转移
    + 买断式回购：债券所有权发生转移
        + 提供了给逆回购方做空市场的机会

+ Fr007：银行间市场7天质押式回购利率，是国内常见的浮动利率

+ 在其发展过程中，互换市场形成的一些*运作机制*也在很大程度上促进了该市场的发展。

#### 互换的运作机制

##### 1. 做市商制度（Market Maker）
+ 互换：OTC产品

    + 互换市场的做市商被称为互换交易商或互换银行（Swap Bank）

+ 利率互换市场的做市商制度尤其发达：

    + 利率互换的同质性较强

    + 利率风险的套保比较容易

    + 协议主文
    + 附件（ Schedule）

##### 3. 其他市场惯例
+ 浮动利率的选择
    +  LIBOR
    +  RMB: 7天回购利率. SHIBOR. 一年期存款利率和一年期贷款利率
    + 一般来说，浮动利率的确定日为每次支付日的前两个营业日或另行约定
+ 净额结算
    + 利率互换在实际结算时通常尽可能地使用利息净额交割，从而使得本金成为名义本金。
    + 净额结算能很大地降低交易双方的风险敞口头寸，从而降低信用风险

+ 互换行情

    + 互换多头（Long）和空头（Short）：

    + 与FRA相同，市场通常将利率互换交易中*固定利率的支付者*（fixed rate payer）称为互换买方，或互换多头；

    + 而将*固定利率的收取者*（fixed rate receiver）称为互换卖方，或互换空头。

+ 买价（Bid Rate）和卖价（Ask Rate）

    + 在做市商制度下，做市商每天都会进行双边互换报价，买价就是做市商在互换中收到浮动利率时*愿意支付的固定利率*；卖价则是做市商在互换中支付浮动利率时*要求收到的固定利率*。
    
    + 显然卖价应高于买价，否则做市商没有差价可赚。

    + 买价与卖价的算术平均为*中间价（middle rate）*，就是通常所说的互换利率（swap rate）

    + 定盘曲线 & 收盘曲线
        + 定盘曲线采样时间是中午，收盘曲线采样时间是下午收盘时间段

+ 互换头寸的结清
    + 出售原互换协议
        + 需要征求对手方同意（因为存在信用风险）

    + 对冲原互换协议（和另一个对手方签相反的互换合约）
        + 如果其中一个对手方违约，另一个合约仍要继续

    + 解除原有的互换协议
        + 需要计算互换合约现在的价值

+ 与远期合约相似，利率互换的定价有两种情形：
    1. 在协议签订后的互换定价，是根据协议内容与市场利率水平确定利率互换合约的价值，可能为正，也可能为负。

    2. 在协议签订时，一个公平的利率互换协议应使得双方的互换价值相等。因此协议签订时的互换定价，就是选择一个使得互换的初始价值为零的互换利率
        + t=0时，互换合约价值为0

##### 债券组合定价法
+ 互换多头：$V_{long} = B_{float} - B_{fix} $

+ 互换空头：$V_{long} = -B_{float} + B_{fix} $

+ 已知债券定价方法：

    + $B_{fix} = \Sigma_{i=1}^{n}CF_{i}* e^{-r_{i}*t_{i}} $
        + $r_{i}$ 为每一次计息时刻 $t_{i}$ 的到期收益率，此时的YTM由浮动利率决定

    + $B_{float} = (A+k^{*})*e^{-r_{1}t_{1}} $
        + $ k^{*} $ 为下一次付息的利息，$r_{1}$ 为这一期期初的浮动利率，*贴现率等于浮动利率*
        + 选定证券存续期内的任一时点， 证券的价值等于该时刻的证券价值加上现在到该时点之间现金流的贴现值
            + 因为只能知道下一次付息的利息，后面的信息都不知道 
        + 在每一期重定利率的时间点（利息已经付完），债券价格正好等于面值
        + 在发行时间点(t=0)，债券价格也等于面值

        


##### FRA定价法
+ $V_{long} = A(e^{r_{F}(T^{*}-T)}-e^{r_{K}(T^{*}-T)})e^{-r^{*}(T^{*}-T)} $
    + $ V_{short} = -V_{long} $
    + 从利率期限结构中估计出FRA对应的远期利率，即可得到每笔FRA的价值，加总即为利率互换多头的价值
        + 对每一期现金互换求FRA价值，然后求和
    + $r_{F}$ 是由即期浮动利率估计的远期利率，$r_{K}$ 为固定利率，$r^{*}$ 为贴现率（等于即期浮动利率）
        + $ r_{F} = \frac{[r^{*}(T^{*}-t) - r(T-t)]}{(T^{*}-T)}$

##### 确定合理的互换利率（固定利率）
+ 合理的互换利率就是使得*利率互换价值为零*的固定利率
    + 我们所求的固定利率是*息票率*而非贴现率
    + 债券组合法：$B_{fix} = B_{float}$
        + 若浮动利率债券是新发行的，则 $B_{fix} = B_{float} = A_{B_{float}} $，固定利率债券价值等于浮动利率债券的本金

##### 互换利率
+ 利率互换协议中合理的固定利率就是使得互换价值为零的利率水平，也就是通常所说的互换利率。
+ 现实中的互换利率是市场以一定的计息频率为基础、就特定期限形成的互换中间利率。

##### 互换收益率曲线
+ 固定利率债券的票面利率就是互换利率，浮动利率债券的利率就是SHIBOR。因此，互换利率与SHIBOR存在内在的联系
    + 加上所覆盖的时间期限(隔夜~1年)不同，SHIBOR与互换利率常常互相补充，形成市场中所称的“互换利率期限结构”或“互换收益率曲线”
+ 一般来说，1年以下的短期期限使用SHIBOR，中长期期限则使用*互换利率中隐含的即期利率*
    + 因为此时互换中的固定利率债券不一定是平价债券，故互换利率（息票率）不一定等于零息债券的到期收益率
    + 互换利率实际上是附息票债券的平价到期收益率，并非即期利率，故此在构造互换利率期限结构时，需要从平价到期收益率信息中提取出相应期限的即期利率
    + 所谓平价到期收益率，是指在当前即期利率期限结构下，使得附息票债券价格等于面值的票面利率
        + 对于0~T期的互换而言：$ CF_{0}e^{-r_{0}*0}+CF_{0}e^{-r_{1}*1}+...+CF_{T}e^{-r_{T}*T} = A_{B_{float}} $,通过解出 $r_{T}$,即可得到此时的即期浮动利率
    


+ 互换曲线不是无风险的利率期限结构，但与国债利率期限结构相比，互换曲线有其自身的优势：
    + 新的互换*会在市场上不断产生*，这使得特定期限的互换利率具有*延续性*，几乎每天都可以估计出特定期限的互换利率；而特定期限的国债利率则往往只有在国债新发行之后才能更像，以发行期限进行循环。
    + 互换是零成本合约，其*供给是无限*的，不受发行量的制约和影响；而国债由于供给的制约会产生新券（on the run）和旧券（off the run）的价格差异。
    + 对于许多银行间的金融衍生产品来说，与无风险利率相比，互换利率由于反映了其现金流的*信用风险*与*流动性风险*，是一个更好的贴现率基准

+ 无风险利率期限结构是根据国债价格信息构造的，互换收益率曲线则主要基于银行间市场报价形成的，这两者的差异被称为“互换价差（互换利差）”（swap spread），主要体现了*银行间市场的信用风险、流动性风险与税收*等的差异
    + 显然，互换收益率高于国债收益率

#### 货币互换的定价

+ 货币互换的现金流
    1. 期初交换本金
    2. 期间交换利息
    3. 期末交换本金

##### 运用债券组合为货币互换定价

+ 对于收入本币利息、付出外币利息的那一方：
    + $V^{D-long}_{swap} = B_{D}-S_{0}B_{F} $
        + $S_{0}$ 为当前即期汇率（直接标价：N外币/本币）
        + $B_{D}$ 为本币债券，$B_{F}$ 为外币债券
+ 对于付出本币利息、收入外币利息的那一方:
    + $V^{D-short}_{swap} = -V^{long}_{swap} = S_{0}B_{F} - B_{D} $

##### 运用FXA组合为货币互换定价

+ 计算并加总货币互换中分解出来的每笔FXA的价值，就可得到相应货币互换的价值
+ 对于每一笔FXA:$ f = (A*F-D)e^{-r(T-t)} $
    + A为外币利息；D为本币利息。
    + 此时是收外币利息，付本币利息

#### 互换的风险

+ 与互换相联系的风险主要包括：
    + 信用风险
        + OTC交易的信用风险（对手方风险）
            + 当利率或汇率等市场价格的变动使得互换对交易者而言价值为正时存在
        + 利率互换
            + 交换的仅是*利息差额*，其交易双方真正面临的信用风险暴露远比互换的名义本金要少得多
        + 货币互换
            + 由于进行*本金*的交换，其交易双方面临的信用风险比利率互换要大一些
    + 市场风险
        + 与互换相联系的市场风险主要可分为利率风险和汇率风险：
            + 对于利率互换来说，主要的市场风险是利率风险
            + 对于货币互换而言，市场风险包括利率风险和汇率风险
        + 利率风险的管理：

    + 净额结算
        + 减小风险敞口

    + 信用衍生产品
        + CDS

##### 期权的定义
+ 期权是买卖未来权利的合约，赋予其购买者（权利方）在规定期限内按双方约定的价格（行权价）购买或出售一定数量某种标的资产的权利。
+ 协议要素
    + 买卖双方
        + 谁有“权利”谁就是多头
    + 约定的权利
    + 行权期限
    + 行权价（对比远期：交割价）
    + 交易数量
    + 期权价格（期权费）
+ 在远期里面，$payoff = profit$，因为期初无成本；期权里面存在期权费，所以$ payoff = profit +  option \space cost $

+ 期权买者(多头)的权利划分：
    + 看涨期权（Call Option，认沽期权）：认购期权（或买权），赋予期权买者未来按约定价格购买标的资产的权利
    + 看跌期权（Put Option，认购期权）：认沽期权（或卖权），赋予期权买者未来按约定价格出售标的资产的权利

+ 本质上是一种金融合约，是买卖双方关于未来某种权利的协议。
    + 多头：支付期权费后，只有权利，没有义务
    + 空头：收取期权费后，只有义务，没有权利
        + 存在违约风险，需要交保证金

    seller & 按K价格卖出的义务 & 按K价格买入的义务

#### 期权市场

1. 合约单位
    + 合约单位，也被称为交易单位或合约规模，就是一张期权合约所包含的标的资产数量。
        + 股票期权：100股股票
        + 股票指数期权：标的指数点数乘以每点100
        + 期货期权：一张标的期货合约
        + 外汇期权：视交易所不同和货币种类不同而不同
    + 50ETF期权：*10000*份上证50ETF

2. 行权价格
    + 合约中的行权价格由交易所事先设定。
        + 交易所准备上市某种期权合约时，首先根据该合约标的资产的最近收盘价，依据某一特定的形式来确定一个*中心行权价格*，然后再根据特定的幅度设定该中心价格的上下各若干级距（Intervals）的执行价格。
        
    + 在期权合约规格中，交易所通常只规定行权价格的级距。
        + 合约标的价格发生变化，导致已挂牌合约中的虚值或实值合约数量不足时，交易所将依据行权价格级距，依序加挂新行权价格的合约。

+ 平值期权（In the money）、虚值期权（Out of the money）和平价期权（At the money）：市场惯例
    + 不考虑时间因素，期权的价值取决于标的资产市价（S）与行权价格（K）的差距。
    + 现实中，行权价最接近现货价的期权也可以视为平价期权
    + 看涨期权
        + $S>K$ 时，实值期权
        + $S=K$ 时，平价期权
        + $S<K$ 时，虚值期权
        + 行权价K越低，看涨期权越值钱
            + 实值期权永远更贵
    + 看跌期权
        + $K>S$ 时，实值期权
        + $K=S$ 时，平价期权
        + $K<S$ 时，虚值期权
        + 行权价K越高，看跌期权越值钱
            + 实值期权永远更贵

3. 除权除息
    + 国内的场外期权受红利保护：如果公司派发现金红利，除权日后行权价格要相应调整。

    + 发放红利之后股票除权：$S^{'}_{T} = S_{T}-I_{T}$

    + 境外交易所流通的期权通常不受红利保护，但在股票分割或送红股（用股票发红利）时要调整（国外的惯例）
        + 对m（即m股股票分割为n股）股票分割之后，行权价格降为原来行权价格的m/n，每一期权合约所包含的标的资产数量上升到原来的n/m倍。
            + 0送10”：$ K' = K*\frac{10}{10+10} $
            + 股票分割为n股:$ K' = K*\frac{m}{n} $
        + 发放n%的股票红利等同于$100+n$对0的分割。

    + 上证50ETF期权受红利保护(国内的惯例)
        + 原则：合约标的总价值不变；行权总价值不变
        + $K^{'}_{T} = K_{T}-I_{T} $
        + P155：上交所除权除息规定（课外阅读，公式不作要求）

4. 行权方式
    + 行权的比例要比期货交割比例高得多
        + 一般选择平仓，而不是交割
    + 现货期权的行权：
        + 以行权价对标的资产进行实际的交收。
    + 指数期权的行权：
        + 行权价与期权行权日当天交易结束时的市场价格之差以现金进行结算。
    + 期货期权的行权：
        + 方行权时，将从期权义务方处获得标的期货合约的相应头寸，再加上行权价与期货结算价之间的现金差额。
    + 50ETF期权：实物交割

#### 期权交易机制

1. 持仓限额和行权限额
    + 头寸限额（Position Limit）：每个投资者在市场的一方所能持有的期权头寸的最大限额
        + 多头/看跌空头
        + 空头/看跌多头

    + 行权限额（Exercise Limit）：一个期权权利方在规定的一段时间内所能行权的期权合约的最大限额
    + 计算标准：合约数量/合约总金额

    + 交易所规定期货期权中期权头寸与相应的期货头寸合并计算
        + 如果都认为标的资产会涨，期货多头也认为标的资产会涨
        
    + P158：上证50ETF期权持仓限额（课外阅读，不作为考试要求）

2. 买卖指令和委托指令
    + 交易指令：上交所
        + 买入建仓，即买入一个期权，建立一个新头寸；
        + 卖出建仓，即卖出一个期权，建立一个新头寸；
        + 买入平仓，即买入一个期权，对冲原有的空头头寸；
        + 卖出平仓，即卖出一个期权，对冲原有的多头头寸；

        + 开仓，是指投资者在持有足额标的证券的基础上，卖出相应数量的认购期权合约。
            + 已经有现货ETF，可以把现货锁定之后卖空期权
            
        + 平仓，是投资者持有备兑持仓头寸时，申请买入相应期权将备兑头寸平仓的指令
            + 已经有抵押的现货，所以空头的违约风险较低

    + 对未平仓合约的影响
        + 和期货一样
    + P158：上交所委托指令（课外阅读，不作为考试要求）

3. 保证金制度
    + 多头
        + 交了期权费，无需缴纳保证金
            + 只有权利，没有义务，不存在违约可能性

    + 空头
        + 期货保证金，分为初始保证金和维持保证金
        + 保证金取决于期权种类和市场状况
            + 空头：抵押品为ETF现货
            + 空头：抵押品为K现金

    + 100%保证金要求： Covered Options
        + 100%标的资产S（call）或100%资金K（put）

    + 100%保证金（eg. 国外股票期权）
        + 初始保证金：两式取较大者$max(m_1,m_2)$
            + 期权的期权费收入加上期权标的资产价值的20%减去期权处于虚值状态的数额（如果有这一项的话）
                + $m_1 = 期权费 + S_t*20\% - 虚值部分 $
                + 为什么扣除的是虚值：虚值期权的空头违约风险较低，不需要太高保证金

            + 期权的期权费收入加上标的资产价值（看涨期权）或执行价格（看跌期权）的10%
                + $ m_2 = 期权费+ 10\%*max(S_{t},K) $
            + 保证金为二者取max的结果，因为虚值有可能特别大，出现保证金为负

        + 维持保证金
    
    + 保证金：上证50ETF期权
        + $认购期权义务仓开仓保证金＝[合约前结算价+Max(12\%×合约标的前收盘价-认购期权虚值,7\%×合约标的前收盘价)]×合约单位$
        + $认沽期权义务仓开仓保证金＝Min[合约前结算价+Max(12\%×合约标的前收盘价-认沽期权虚值，7\%×行权价),行权价]×合约单位$
            + 保证金为K时，为100%安全保证金
        + 虚值：
            + $认购期权虚值＝Max（行权价-合约标的前收盘价，0）$
            + $认沽期权虚值＝Max（合约标的前收盘价-行权价，0）$

    + 期权头寸的保证金

#### 期权的回报与盈亏分布

```python
import matplotlib.pyplot as plt
import numpy as np
S_T = np.arange(start=0, stop=100,step=0.1)
K = 50
c,p = 10,10
call_long = np.where(S_T>K,S_T-K,0)
put_long = np.where(S_T<K,K-S_T,0)
call_short = np.where(S_T>K,K-S_T,0)
put_short = np.where(S_T<K,S_T-K,0)

ax_1 = plt.subplot(2,2,1)
ax_2 = plt.subplot(2,2,2)
ax_3 = plt.subplot(2,2,3)
ax_4 = plt.subplot(2,2,4)

ax_1.plot(S_T,call_long,S_T,call_long-c)
ax_1.set_title('看涨多头:回报&盈亏')
ax_2.plot(S_T,put_long,S_T,put_long-p)
ax_2.set_title('看跌多头:回报&盈亏')
ax_3.plot(S_T,call_short,S_T,call_short+c)
ax_3.set_title('看涨空头:回报&盈亏')
ax_4.plot(S_T,put_long,S_T,put_long+p)
ax_4.set_title('看跌空头:回报&盈亏')
```

+ 以上四幅图并没有给出回报和盈亏对应的概率分布，容易让人误解期权多空双方在盈亏上严重不对称。实际上期权多头的特征是“大概率亏期权费，小概率获利”；期权空头则是“大概率赚期权费，小概率亏损”。
  + 在一个公平市场上，期权多空双方的期望损益是相对称的。
+ 只要在卖出期权后，能有效地对冲小概率亏损的风险，仍然是可以充当期权空方的，如专业的金融机构。但如果单纯卖出期权，就是在冒小概率的巨亏风险，是高风险的行为。

call_{short} & -max(S_{T}-K,0) & -max(S_{T}-K,0)+c_{t} \\

\end{matrix}

#### 期权价格的特性

##### 期权价格：内在价值与时间价值
+ 期权价格（或价值） =内在价值 +时间价值
    + 内在价值（Intrinsic Value）是指在*不考虑标的资产价格波动*的情况下，期权条款*赋予期权多头的最高价值*；（或有回报的最大值）
    + 时间价值(Time Value）是指在期权剩余期限内，*标的资产价格的波动*为期权多头带来收益的可能性所*隐含的价值*（波动带来的价值）。
        + 时间价值会受内在价值的影响，但内在价值不受时间价值的影响。

##### 内在价值与平值点(本书定义与市场不同)
+ 市场惯例：将当前的标的资产价格$S_{t}$作为所有期权的平值点，并未区分美式期权与欧式期权的不同特点。
    + call: max(S_{T}-K,0)$
    + put: max(K-S_{T},0)$
    + 平值点：$ K_{ATM} = S_{T} $
    + 为什么不用市场惯例？
        + 为了保证期权的时间价值不会小于0，所有期权价格的下限就是其内在价值，而且期权的时间价值都在平值点最大
1. 欧式看涨期权的内在价值
    + 不考虑时间价值， 欧式看跌期权合约与远期合约空头的唯一区别就是前者只有权利没有义务
        + $max(-f,0) = max[(F-K)e^{e(T-t)},0] $
            + 内在价值为负时，可以选择不行权
        + 此时的平值点(At The Money)：$ K_{ATM} = F $(和红利支付方式有关)
2. 欧式看跌期权的内在价值

##### 实值期权、平值期权与虚值期权
+ 平值期权（At the Money, ATM）
    + 平值点（$K_{ATM}$）就是期权内在价值由正值变化到零的行权价格的临界点
+ 实值期权（In the Money, ITM）
    + Call：行权价格<平值点
    + Put:行权价格>平值点
+ 虚值期权（Out of the Money, OTM）
    + Call：行权价格>平值点
    + Put：行权价格<平值点
+ 虚值和平值期权的内在价值为零，其全部价值都体现为时间价值；实值期权具有正的内在价值。

##### 在值程度
+ 与实值期权和虚值期权密切相关的概念是期权的在值程度（moneyness），其中一种定义：
    oneyness = ln(\frac{K}{K_{ATM}}) $
+ 平值期权的在值程度为0
+ 实值call和虚值put的在值程度小于0
+ 虚值call和实值put的在值程度大于0

##### 期权的时间价值
+ 期权的时间价值是指在期权剩余期限内，标的资产价格的波动为期权多头带来收益的可能性所隐含的价值（波动带来的价值）。
    + 期权的时间价值是基于期权多头权利义务不对称这一特性，在期权到期前，标的资产价格的变化可能给期权多头带来的收益的一种反映。
    书对期权内在价值的定义，期权的时间价值一定大于等于0。
    
+ 影响期权时间价值的因素：
    + $T-t$:其他条件相同的情况下，一般来说，距离期权到期时间越长，期权的时间价值越大，对美式期权来说尤其如此。
    + $\sigma$:其他条件相同的情况下，标的资产价格的变化（波动）越大，期权的时间价值就越大。
    + 在值程度：按本书的平值点和在值程度定义，*平值期权*（在值程度为0,$K=K_{ATM}$）的时间价值最大，实值和虚值期权的时间价值则随着实值和虚值程度的提高而逐渐递减。

##### 期权价格的影响因素
+ 标的资产的市场价格
    + 希腊字母：$\Delta = \frac{\partial V_{t}}{\partial S_{t}}$
    + $S_{t}$越大，$c^{E}_{t}$ 越大， $p^{E}_{t}$越小
+ 行权价格
    + $K$越大，$c_{t}$ 越小， $p^{E}_{t}$越大
+ 剩余期限
    + $T-t$越大，$c^{A}_{t}$ 越小（行权时间区间更大），一般而言 $c^{E}_{t}$ 越小
+ 标的资产价格的波动率
    + $\sigma$越大，$c_{t}$、$p^{E}_{t}$越大
    + 如果没有波动，则期权就没有时间价值
        + 此时看涨期权仍有行权价值，而看跌期权价值为0
+ 无风险利率 （理论上不能确定）
    + 利率r越高，标的资产的预期收益率也越高，这意味着对应于标的资产现在特定的市价$S_{t}$，未来预期价格 E(S_{T})越高。（明显）
    由于贴现率r越髙，未来同样预期回报$(K-S_{T})^{+}$的现值就越低。（不明显） 
    + 看跌期权价值 $p^{E}_{t}$ 会降低，看涨期权价值 $c^{E}_{t}$ 会上升
+ 标的资产的收益


##### 不提前行权无红利资产美式看涨期权的合理性
+ 对于无红利资产美式看涨期权来说，$T_{0}$ 时刻提前行权对权利方的影响：
    + 多头需提前支付行权价，因此要多损失行权价的利息 $K*e^{r(T-T_{0})}$；
    + 多头提前获得标的资产，但该标的资产没有红利，因此*提前获得标的资产并没有好处* $S_{T_{0}}*e^{\delta(T-T_{0})}=0$；
    + 提前行权意味着失去了*期权的时间价值*，而时间价值永远大于等于零。

+ 提前行权无红利资产的美式看涨期权是不明智的。
    + 对于无红利资产的美式看涨期权，如果预判未来资产价格可能下跌，最优的选择是*卖掉期权*，而不是提前行权。
        + 只有如同雇员股票期权这种无法转卖的期权，才会在实值的时候提前行权
    + 在其他条件相同的情况下，无红利资产美式看涨期权和欧式看涨期权实际上是等价的，它们的价值是相同的，其内在价值和时间价值都相等。

##### 提前行权无红利资产美式看跌期权的合理性
+ 对于无红利资产美式Put来说，$T_{0}$ 时刻提前行权对权利方的影响：
    + 多头提前得到行权价，因此可以获得行权价的利息 $K*e^{r(T-T_{0})}>0$；
    + 多头提前支付标的资产，但该标的资产没有红利，因此提前行权并没有损失红利 $S_{T_{0}}*e^{\delta(T-T_{0})}=0$；
    + 提前行权意味着失去了期权的时间价值。

+ 是否提前行权无红利资产的美式看跌期权，主要取决于期权的实值程度、无风险利率水平等因素。
    + 一般来说，只有当实值程度较高(可以抵消损失的时间价值)、利率较高时，提前行权无红利资产美式看跌期权才可能是有利的。

##### 提前行权有红利资产美式看涨期权的合理性
+ 提前行权的好处：
    + 获得红利
+ 坏处：
    + 提前支付行权价格，会损失行权价的利息
    + 浪费期权的时间价值

+ 如果红利小于行权价格的利息$I<K*e^{r(T-T_{0})}$，那么提前行权是不合理的；但如果红利大于行权价格的利息，那么提前行权就*有可能*是合理的。
    + 一般来说，只有当实值程度较高、红利较高、利率较低时，提前行权有红利资产美式看涨期权才可能是有利的。
    行权的得到的红利I应当要弥补利息 $K*e^{r(T-T_{0})}$ 与时间价值，才要提前行权

##### 提前执行有红利资产美式看跌期权的合理性
+ 提前行权有红利资产的美式看跌期权意味着自己放弃标的资产的红利
    + 与无红利资产的美式看跌期权相比，有红利资产美式看跌期权提前行权的可能性减小，但仍无法完全排除提前行权的可能性

##### 期权价格的上限
+ 欧式看涨：
\begin{cases}
c^{E}_{t} \leq S_{t} , 无红利资产 \\
c^{E}_{t} \leq S_{t} -I ， 有红利资产
\end{cases}

+ 美式看涨：
$$c^{A}_{t} \leq S_{t}$$
+ 对于看涨期权，回报的最大值就是得到1单位标的资产（去除红利）
+ 欧式看跌：
\begin{cases}
c^{E}_{t} \leq K*e^{r(T-t)} , 无红利资产 \\
c^{E}_{t} \leq K*e^{r(T-t)} ， 有红利资产
\end{cases}
+ 美式看跌
$$c^{A}_{t} \leq K$$
+ 对于看跌期权，回报的最大值就是得到K单位现金（加上持有资产的红利）

+ 期权价格的下限：
  + 按照本书内在价值的定义，由于期权的时间价值一定大于0，因此期权价格一定大于等于期权的内在价值，即*期权的内在价值*就是期权的下限。

### 第十一章 B-S-M期权定价模型

#### 股票价格的变化过程

##### Markov过程
+ 马尔可夫随机过程，是指只有变量的当前值才与未来的预测有关，变量过去的历史和变量从过去到现在的演变方式与未来的预测无关
  + t时刻的信息能够真正影响t+1时刻的数学期望
  $E(X_{n+1} = x|X_{1\~n}=...) = E(X_{n+1} = x|X_{n}=x) $

##### 标准布朗运动(维纳过程)
+ 布朗运动：$ dS_{t} = S_{t}*e^{\mu t \sigma W_{t}} $
  + $W_{t}$ 为标准布朗运动
+ 布朗运动的四大特征：
  +  初始值为零 $ W_{0}=0 $
    
  + 独立平稳增量：对于任何两个不重合的时间间隔Δt， Δz的值相互独立 $Cov(P(W_{t+s}-W_{t}=...),P(W_{t}=...))=0$
  + 独立同分布：在任意小的时间间隔 Δt内，增量均服从均值为零、方差等于时间长度的正态分布 $W_{\Delta t} \sim N(0,\Delta t)$

+ 使用布朗运动 P199
  + 分布：经验事实证明，股票价格的连续复利收益率*近似地服从正态分布*（P204图11.1）

  + 可以证明，标准布朗运动（维纳过程）是一个*马尔可夫随机过程*，从而与弱式EMH相符

  + 布朗运动（维纳过程）在数学上*对时间处处不可导*和二次变分（Quadratic Variation,可以用方差来类比）不为零的性质，与股票收益率在时间上存在转折尖点等性质也是相符的



##### 普通布朗运动：标准布朗运动的扩展（一般化的维纳过程）
+ 普通布朗运动的变量 $x_t = \mu dt +  \sigma dz_{t}$ 是关于时间和 $dz_t$ 的动态过程，$\mu$ 和 $\sigma$ 均为常数
    + $\mu dt$为确定项，漂移率 $\mu$ 意味着x的漂移速度是每单位时间为 $\mu$；
    + $\sigma dz_{t}$ 是随机项，代表着对 $x_t$ 的时间趋势过程所添加的噪声，使变量$x_t$围绕着确定趋势上下随机波动，且这种噪声是由维纳过程的 $\sigma$ 倍给出的， $\sigma^{2}$ 称为方差率，$\sigma$ 称为波动率。

##### $it\hat{o}$过程
+ 最常见的一类伊藤过程，是将at和bt设定为变量xt和时间t的确定性函数：$$ dx_t = \mu(x_{t},t) dt +  \sigma(x_{t},t) dz_t $$
    + $\mu(x_{t},t)$ 与 $\sigma(x_{t},t)$ 为关于t的可料过程
    + 称之为“扩散过程”（diffusion process）

##### 几何布朗运动
布朗运动（ Geometric Brownian Motion）:$$ dS_{t} = S_{t}(\mu dt +  \sigma dz_t) $$
  ，μ (漂移率)和σ (波动率)均为常数
  这是一个漂移项为$μSt$，方差项为$σ^2 S^2_t$的伊藤（扩散）过程
用几何布朗运动来描述股票价格的随机过程。

+ 为什么不用普通布朗运动？
  1. 可以避免股票价格为负从而与有限责任相矛盾的问题
  2. 几何布朗运动意味着股票连续复利收益率服从正态分布，这与实际较为吻合
    复利收益率（不一定为年化收益率）：$$(lnS_{T}-lnS_{t}) \sim N[(\mu-\frac{1}{2}\sigma^{2})(T-t),(\sigma \sqrt{T-t})^{2}] $$
  3. 股票价格的对数服从普通布朗运动，特定时刻的股票价格$S_T$服从对数正态分布（取对数之后为正态）
    $$E(S_{T}) = S_{t}*e^{\mu(T-t)}$$
    $$Var(S_{T}) = S_{t}^{2}(e^{\sigma^{2}(T-t)}-1)e^{\mu(T-t)}$$
    ， μ是Δt时间内股票价格*百分比*的年化预期收益率(对数收益率则是$\mu - \frac{1}{2}\sigma^{2}$)

##### 伊藤引理($it\hat{o}\space lemma$)
伊藤过程：$dx_t = \mu(x_{t},t) dt \sigma(x_{t},t) dz_t$
对于函数$ f(x_t,t) $,有：
$$ df = \frac{\partial f}{\partial t} dt + \frac{\partial f}{\partial x_{t}}dx_t +\frac{1}{2} \frac{\partial^{2} f}{\partial x_{t}^{2}}(dx_{t})^{2} $$ 
  ， $dz_t$ 是一个标准布朗运动
  是上述导数都存在并连续

+ 伊藤引理的推导
  + 我们知道 $x_t$遵循伊藤过程，通过Itô-Doeblin引理可以推导出 $f(x_t, t)$遵循的随机过程。
    1. 泰勒展开：$f(x_t, t) = \frac{\partial f}{\partial t}dt + \frac{\partial f}{\partial x_{t}} + \frac{1}{2} \frac{\partial^{2} f}{\partial x_t^{2}}(dx_{t})^{2} ....$
    2. 注意到$d(x_{t})^{2} = \sigma^{2}dt $，存在dt的同阶无穷小，所以不应该舍去这一项

+ 伊藤引理的应用(对于 $dS_{t} = S_{t}(\mu dt + \sigma dz_{t}) $):
  
  1. 连续复利收益率 $lnS_t$ 的SDE:$ d(lnS_{t}) =  \frac{1}{S_{t}} dS_{t} - \frac{1}{2} (dS_t)^{2} = (\mu - \frac{1}{2} \sigma^{2})dt +\sigma dz_t  $
  
  2. 远期价格 $F_t = S_{t}*e^{r(T-t)}$的SDE: $dF_t = F_{t}[(\mu - r)dt + \sigma dz_t] $

衍生产品价格是标的资产价格和时间的函数，因此Itô-Doeblin引理在衍生产品分析中扮演重要的角色。

##### 百分比收益率 & 对数收益率

+ 在一个很短的瞬间Δt，股票价格的百分比收益率服从：$ \frac{\Delta S}{S} = \mu \Delta t+ \sigma * \epsilon \sqrt{\Delta t} $
    + 即在很短的时间内，百分比收益率具有正态分布特征，其均值为$μΔt$，方差为$σ^2 Δt$。

+ 但由于较长时间内的百分比毛收益率为每个短时间百分比收益率的乘积，而正态分布变量的乘积并不服从正态分布，所以几何布朗运动只意味着短时间内的百分比收益率服从正态分布，在长时期内其正态分布的性质不再存在。

+ 而较长时间的连续复利收益率为每个短时间连续复利收益率之和，独立正态分布变量之和仍为正态分布，因此连续复利收益率始终服从正态分布。
  + $d(lnS_{t}) = (\mu - \frac{1}{2}\sigma^{2})dt + \sigma *dz_{t}$

+ 短时间间隔Δt内百分比收益率的年化期望值为μ，而连续复利收益率的年化期望
值仅为(μ−σ2/2)，股票价格的波动越大，两者的差异也越大

+ T时刻股票价格ST的期望值：
  + 由于$e^x$函数是非线性的，$E(e^{x}) \neq e^{E(x)}$
  + 因此股票价格的期望值$E_{P}[S_T]=S_t*e^{μ(T-t)}$并不意味着年化连续复利收益率的期望值为μ

##### 鞅(Martingale)
+ 定义：$$ E(X_{T}|F_{t}) = X_{t} $$
+ 判断方法：$$ dX_{t} = \sigma(X_{t},t)dz_{t} + 0dt $$

#### BSM期权定价公式（无风险定价法）

##### 假设
+ 不存在无风险套利机会；(无套利)
+ 允许卖空标的股票；
+ 没有交易费用和税收
+ 证券交易是连续的，价格变动也是连续的；
+ 所有证券都完全可分；（可复制）
+ 股票价格遵循几何布朗运动，即μ和σ为常数；
+ 在衍生产品有效期内，无风险连续复利利率r为常数；
+ 在衍生产品有效期内标的股票无红利。（股价连续，无“跳跃”现象）

##### B-S的PDE

1. $f$为依赖于$S_{t}$的衍生品价格，根据伊藤引理，可得：$$ df_{t} = \frac{\partial f }{\partial t}dt + \frac{\partial f}{\partial S}dS +\frac{1}{2}\frac{\partial^{2} f}{\partial S^{2}}dS $$
2. 整理可得：$$ df_{t} = (\frac{\partial f }{\partial t}+ \frac{1}{2}\sigma^{2}\frac{\partial^{2} f}{\partial S^{2}})dt + \frac{\partial f}{\partial S}dS  $$
3. 由于$dS$中存在随机过程$dz_{t}$，因此考虑使用无风险组合$\pi_{t} = f_{t} - \frac{\partial f}{\partial S}*S$来对冲风险
4. 可以写出$pi_{t}$的SDE:$ (\frac{\partial f }{\partial t}+ \frac{1}{2}\sigma^{2}\frac{\partial^{2} f}{\partial S^{2}})dt $
5. 由于$pi_{t}$已经成为无风险组合，其收益率为：$ d\pi_{t} = \pi_{t}*r dt $
6. 因此，可以列出方程：$$ \frac{\partial f }{\partial t}+ \frac{1}{2}\sigma^{2}*\frac{\partial^{2} f}{\partial S^{2}} + r*S*\frac{\partial f}{\partial S} - r*f =0 $$

+ 在一定的边界条件下求解BSM偏微分方程，就可以得到ft所满足的表达式，也就是衍生证券的定价公式。
  + 注：对于不同衍生品而言，其主要的边界条件就是到期回报。虽然同一标的的衍生证券所满足的偏微分方程相同，但不同边界条件得到的解是不同的。
  + 注：当$S_t$和t变化时，$\frac{\partial f}{\partial S}$ 的值也会变化，因此上述组合的价值并不是永远无风险的，它只是在一个很短的时间间隔 Δt中才是无风险的。

##### 风险中性定价原理
+ 观察BSM偏微分方程可以发现，受制于主观风险收益偏好的标的证券预期收益率μ并未包括在衍生证券的价值决定公式中。这意味着，无论风险收益偏好状态如何，都不会对f的值产生影响。
    + $\mu$体现了投资者的风险爱好程度
+ 因此我们可以作出一个可以大大简化我们工作的技术假定：在对衍生证券定价时，所有投资者都是风险中性的。

##### 风险中性世界（Q测度）
+ 风险中性世界中，投资者只关心资产的预期回报（Expected Payoff），而不关心该回报的风险。这样预期回报相等的资产，其目前的价格都相等。
+ 对于到期回报为1的贴现式债券而言：$E_{Q}(1*B_{T}|\Bbb{F}_{t}) = e^{r(T-t)}$
+ 对于到期预期回报为$E(S_T)$的资产而言，其价格为$E_{Q}(S_{T}|\Bbb{F}_{t}) = E(S_T)*e^{-r(T-t)} $

+ 在所有投资者都是风险中性的条件下（有时我们称之为进入了一个“风险中性世界”）：
  + 所有可交易证券的漂移率(预期收益率)都等于无风险利率r。
  + 所有现金流的期望值都应该使用无风险利率r进行贴现。
  + 这就是风险中性定价原理。

+ 注：风险中性定价原理来源于BSM偏微分方程中没有μ的存在。
  + 而偏微分方程是在无套利条件下，运用*可交易*的证券复制出无风险组合得到的，因此，*只有可交易证券*在风险中性世界中的预期收益率才为无风险利率 r。

+ 必须强调的是：风险中性定价仅仅是为了衍生品定价而做出的技术假定，并不意味着我们真的认为市场投资者是风险中性的，但通过这种假定得到的定价结果，不仅适用于风险中性的情形，也适用于投资者厌恶风险的所有情形。

+ 运用风险中性定价法的条件：无套利和可复制

##### 无红利资产欧式看涨期权的定价公式

1. 已知T时刻payof：$X_{T} = max(S_{T}-K,0)$
2. 因此t时刻的欧式看涨期权价值$c^{E}_{t}$即为：$$ c^{E}_{t} = e^{-r(T-t)}E_{Q}(X_{T}|\Bbb{F}_{t}) $$
3. 已知$S_{T} = S_{t}*e^{\mu*(T-t)+\sigma*(W_{T}-W_{t})}$
4. 构造风险中性测度：$\tilde{W}_{t} = W_{t}+ \frac{\mu - r + \frac{1}{2}\sigma^{2}}{\sigma}$
5. 代入风险中性期望：
    
    \begin{aligned}
    E_{Q}(X_{T}|\Bbb{F}_{t}) &= E_{Q}(max(S_{T}-K,0)|\Bbb{F}_{t}) \\
    &= E_{Q}(max(S_{t}*e^{\mu*(T-t)+\sigma*(W_{T}-W_{t})}-K,0)|\Bbb{F}_{t}) \\
    & = E_{Q}(max(S_{t}* e^{(r-\frac{1}{2}\sigma^{2}) * (T-t)+\sigma*(\tilde{W}_{T}-\tilde{W}_{t})}-K,0)|\Bbb{F}_{t}) \\
    \end{aligned}
6. 确定积分界限：令$S_{t}* e^{(r-\frac{1}{2}\sigma^{2}) * (T-t)+\sigma*(\tilde{W}_{T}-\tilde{W}_{t})}-K \geq 0$ 
    
    可得：$ \frac{\tilde{W}_{T}-\tilde{W}_{t}}{ \sqrt{T-t}} \geq - \frac{ln(\frac{S_{t}}{K}+(r-\sigma^{2}))}{\sigma \sqrt{T-t}} $
    
    设 $y = \frac{\tilde{W}_{T}-\tilde{W}_{t}}{ \sqrt{T-t}} $,$ d_{2} = \frac{ln(\frac{S_{t}}{K}+(r-\sigma^{2}))}{\sigma \sqrt{T-t}} $,$d_{1} = d_{2}+\sigma\sqrt{T-t} = \frac{ln(\frac{S_{t}}{K}+(r+\sigma^{2}))}{\sigma \sqrt{T-t}}$
7. 积分求期望：

    \begin{aligned}
    E_{Q}(X_{T}|\Bbb{F}_{t})&= \int_{-d_{2}}^{+\infty} S_{t} * e^{(r-\frac{1}{2}\sigma^{2}) * 
    (T-t)+\sigma*\sqrt{T-t}*y} *\frac{1}{\sqrt{2\pi}} *e^{-\frac{y^{2}}{2}} dy - \int_{-d_{2}}^{+\infty} K *\frac{1}{\sqrt{2\pi}} *e^{-\frac{y^{2}}{2}} dy\\
    & = S_{t}* e^{r(T-t)} * \int_{-d_{2}}^{+\infty} \frac{1}{\sqrt{2\pi}}* e^{-\frac{y^{2}}{2}+\sigma* \sqrt{T-t}*y-\frac{1}{2} *\sigma^{2}} dy -K *\Phi(d_2) \\
    &=S_{t}* e^{r(T-t)} * \int_{-d_{2}-\sigma*\sqrt{T-t}}^{+\infty} \frac{1}{\sqrt{2\pi}}* e^{-\frac{y^{2}}{2}+\sigma* \sqrt{T-t}*y-\frac{1}{2} *\sigma^{2}}d(y+ \sigma *\sqrt{T-t})-K *\Phi(d_2)\\
    &= S_{t}* e^{r(T-t)} \Phi(d_1) - K *\Phi(d_2)
    \end{aligned}
8. 因此可得无红利资产欧式看涨期权的定价公式：$$c^{E}_{t} = S_{t}* \Phi(d_1) - K*e^{-r(T-t)} *\Phi(d_2) $$
    + $\Phi(X)$表示标准正态分布的概率分布函数(CDF)

#### 理解B-S-M期权定价公式

##### 复制策略
+ 我们可以用股票和负债复制期权。
  + 可以证明,它是构造无风险组合$\Pi$时的Δ，是复制投资组合中股票的数量，$S_t*N(d_1)$就是股票的市值
  + 而 $K*e^{-r(T-t)}N(d_2)$则是复制交易策略中负债的价值。
+ 由于主要参数都是时变的，因此这种复制策略是动态复制策略，必须不断调整
相关头寸数量

+ 证明：$\frac{\partial V}{\partial S} = \Phi(d_{1})$:
  + $ V_{t} =S_{t}* \Phi(d_1) - K*e^{-r(T-t)} *\Phi(d_2) $
  + 
    \begin{aligned}
    \frac{\partial V}{\partial S} &= \frac{\partial S* \Phi(d_1)- K*e^{-r(T-t)} *\Phi(d_2)}{\partial S} \\
    &= \Phi(d_1) + S*\phi(d_1) * \frac{\partial (d_1)}{\partial S} - K*e^{-r(T-t)} *\phi(d_2) * \frac{\partial (d_2)}{\partial S} \\
    &= \Phi(d_1)
    \end{aligned}
  + 注意到：$ S*\phi(d_1)=  K*e^{-r(T-t)} *\phi(d_2)$,$ \frac{\partial (d_1)}{\partial S} = \frac{\partial (d_2)}{\partial S} $

##### 风险中性概率
+ 可以证明：$N(d_2)$是风险中性测度下$S_T>K$的概率，即欧式看涨期权被行权的*风险中性概率*，因此，$K*e^{-r(T-t)}N(d_2)$可以看成预期行权所需支付现金的现值。
+ 而$E_{Q}(S_{T}|\Bbb{F}_{t}) = E(S_T)*e^{-r(T-t)}$则是在风险中性测度下，一个$S_{t}$如果$S_T > K$就等于$S_{T}$，否则就等于0的变量的期望值， $S_t*N (d_1)$则是这个期望值的贴现值，可以看成期权持有者预期行权所得回报的现值。
+ 因此，整个看涨期权定价公式就是在风险中性测度下期权未来期望回报的现值
