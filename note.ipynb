#### 连续复利（一年计息无数次的利率）
先列出m次计息情况下的终值：
$$A(1+\frac{r_{m}}{m})^{mn}$$
利用高数里面的重要极限：
$$\lim_{x→\infty}(1+\frac{1}{x})^{x} = e$$
整理原式之后，代入极限：
$$(A(1+\frac{1}{\frac{m}{r_{m}}})^{\frac{m}{r_{m}}})^{r_{m}n}  $$
$$A(1+ e^{rn})$$
$A(1+ e^{rn})$表示连续复利情况下的终值

##### 各种计息频次利率与连续利率的转化公式
基础：令终值相等，列等式
$$e^{r} = (1+\frac{r_{m}}{m})^{m} = 1+r_{1}$$
即可推出：
$$r = m ln (1+\frac{r_{m}}{m}) $$

$$r_{m} = m(e^{\frac{r}{m}}-1)$$
尤其是当m = 1时： 
$$r = ln (1+r_{1}) = ln (P_{1})-ln (P_{0})$$

在股票投资场景中，由于对数收益率$ln (\frac{P_{1}}{P_{0}})$比百分比收益率$\frac{P_{1}-P_{0}}{P_{0}}$拥有正态性. 可连加性等优良性质，一般对数收益率更加常用

### 远期&期货概述

#### 远期合约种类
1. 远期*利率*协议（FRA）
    + 定义：买卖双方同意从未来某一商定的时刻开始，在某一特定时期内按协议利率借贷一笔数额确定. 以特定货币表示的名义本金的协议。
    + 通过事先约定利率，借款人可以规避利率上升的风险，贷款人则可以规避利率下跌的风险。无论未来利率升跌，FRA将实际贷款利率锁定为协议利率。
    + $1*4$：一个月后，借款3个月（至第四个月）
2. 远期*外汇*协议
    + 定义：双方约定在将来某一时间按约定的汇率买卖一定金额的某种外汇的合约
    + CNY:在岸市场的远期外汇交易，标的为在岸即期￥汇率；CNH：离岸市场的远期外汇交易，标的为离岸即期￥汇率；NDF:游离于监管之外的外汇市场，标的为在岸市场的￥汇率（CNY. CNH属于可交割的远期，即DF；NDF则无法交割本金,实际结算时使用其他可自由兑换的货币）
    + 买报价（bid）&卖报价（ask）：外汇的买/卖报价是针对银行而言的，银行买外汇时使用买价，卖出时使用卖价

##### 与远期的不同（场内衍生品&场外衍生品的区别）

1. 交易场所不同
    + 远期并没有固定的交易场所， 买卖双方各自寻找合适的对象， 因而是一个无严格组织的分散市场。 
    + 期货合约则在交易所内集中交易， 一般不允许场外交易。 交易所不仅为期货交易提供了交易场所和统一清算， 还为期货交易制定了许多严格的交易规则（ 如保证金制度 .  每日盯市结算制度 . 涨跌停板制度 . 最小价格波动值.  报价方式和最大持仓限额等） ， 并为期货交易提供信用担保。 可以说， 期货市场是一个有组织. 有秩序的统一市场。

2. 标准化程度不同
    + 远期交易遵循“ 契约自由” 的原则， 合约中的相关条件如标的资产的质量.  数量. 交割地点和交割月份都是根据双方具体的需要确定的， 其条款的具体内容往往体现出很强的个性化特征。 
    + 期货合约则是标准化的。 期货交易所为各种期货合约制定了标准化的数量. 质量. 交割地点. 交割时间. 交割方式. 合约规模等条款， 只有价格是在成交时根据市场行情确定的。 

3. 违约风险不同
    + 远期合约的履行主要取决于签约双方的信用， 一旦一方无力或不愿履约， 另一方就会蒙受损失。 目前国际主要远期市场的参与者大多需达到一定的资质， 并通过第三方或资产担保等方式降低远期交易的违约风险。
    + 期货合约的履行则由交易所或清算机构提供担保。 交易双方直接面对的都是交易所， 即使一方违约， 另一方也几乎不会受到影响。 

4.  合约双方关系不同
    + 远期合约是交易双方直接签订的， 而且由于远期合约的违约风险主要取决于交易对手的信用， 因此签约前通常要对交易对手的信誉和实力等方面作充分的了解。
    + 期货合约的履行完全不取决于对方， 只取决于交易所或清算机构， 交易所是所有买方的卖者和所有卖方的买者。 在期货交易中， 交易者根本无须知道对方是谁， 市场信息成本很低。

5. 价格确定方式不同
    + 远期合约的交割价格是由交易双方直接谈判并私下确定的。 
    + 期货交易的价格则是在交易所中通过公开竞价或根据做市商报价确定的。

6. 结算方式不同
    + 远期交易是到期一次性结算的，所以在远期存续期内，实际交割价格始终不变，标的资产市场价格的变化给投资者带来的是账面浮动盈亏，到期结算时标的资产的市场价格与交割价格的差异才是投资者的真实盈亏。

    + 期货则有所不同。由于期货是*每日盯市结算*实现真实盈亏的，因此可以把期货看作一个每日以结算价平仓结清并以该结算价重新开立的合约，每日结算价格就是不断变动的期货交割价格。

    + 由于每天盯市结算，所以期货的最后结算价格仍然是T时刻的真实价格$F_{T}$；远期只进行一次结算，按照买入远期时的价格$F_{0}$交割

7. 结清方式不同
    + 由于远期合约是非标准化的， 不易找到转让对象， 并要征得原交易对手的同意， 因此绝大多数远期合约只能通过到期实物交割或现金结算来结束。
    + 期货合约则可以通过到期交割结算和平仓两种方式结清。 由于交易便利， 期货交易一方无须征得对方同意就可通过平仓来结清自己的头寸并把履约权利和义务转让给第三方。在实际中， 绝大多数期货合约都是通过平仓来了结的。

##### 卖空（Short Selling）

+ 未来需买回归还
+ 此期间需支付原持有者应获得的股利等收入（红利是要还的）

卖空的限制：
+ A股的“卖空”：融券机制，只有融券名单上的资产可以卖空（借券费用很高！）
+ 提升（uptick rules）卖空需要提高报价

1. 无套利方法主要适用于投资性资产，不一定适用于消费性资产
2. 交割价格（Delivery Price）：合约所规定的固定价格 $K$
3. 远期价值：远期合约本身的价值（作为0成本的零和游戏，公平的远期价值应当是0）
4. 远期价格（Forward Price）：使得远期价值为0的合理交割价格 $F$
+ 当 $K = F$时，远期价值为0， $K$ 是一个比较合理的价格

##### 远期价格&期货价格
1. 当无风险利率恒定且对所有到期日都相同时，其他条件相同的远期价格和期货价格相等。
2. 当利率变化无法预测时，两者略有不同 (P48，涉及保证金账户余额的变化)
    + 当标的资产价格与利率正相关时，期货价格高于远期价格
    + 当标的资产价格与利率负相关时，远期价格高于期货价格
3. 远期价格和期货价格的差异幅度还取决于：
    + 合约有效期. 税收. 交易费用. 保证金的处理方式. 违约风险. 流动性等

##### 基本假设
1. 没有交易费用和税收。
2. 市场参与者能以相同的无风险利率借入和贷出资金。
3. 允许现货卖空。
4. 没有违约风险。
5. 当套利机会出现时，市场参与者将参与套利活动，从而使套利机会消失。
    + 我们得到的理论价格就是*没有套利机会下的均衡价格*。
6. 期货合约的保证金账户支付同样的无风险利率。
    + 这意味着任何人均可不花成本地取得远期和期货的多头和空头地位
    + 此时期货价格 = 远期价格

+ 前三条假设可以有条件放宽

##### 复制定价法
+ 基本思路：
    + 构建两种投资组合，令其终值相等，则其现值一定相等(MF:无套利引理 2)；否则就可进行套利，即卖出现值较高的投资组合，买入现值较低的投资组合，并持有到期末，套利者就可赚取无风险收益。

    + 众多套利者这样做的结果，将使较高现值的投资组合价格下降，而较低现值的投资组合价格上升，直至套利机会消失，此时两种组合的现值相等。

    + 这样，我们就可根据两种组合现值相等的关系求出远期价格。

+ 构造投资组合：
    + 组合B：一单位标的资产(无红利资产)
        + 终值：$t = T$时，$W_{T} = S_{T}$
        + 现值：$W_{t} = S_{t}$

    + 组合A：一份远期合约多头 & 一笔数额为$K e^{-r(T-t)}$的现金（无风险投资）

        + 该远期合约规定多头在到期日T可按交割价格K购买一单位标的资产。
        + 终值：$t = T$时,$W_{T} = S_{T}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$（f为合约价值，K为交割价）
    
    + 令组合A. B现值相等，可得：$f = S_{t} - K e^{-r(T-t)}$

+ 无红利资产的现货－远期平价定理（spot-forward parity theorem），或称现货－期货平价定理（spot-futures parity theorem）
    + F为使合约价值f为0的交割价K
    + $f=0$时，可得$K= F = S e^{r(T-t)}$

+ 证明（反证法）
    + 若$K>S e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）:

        \begin{array}{c|c|c|c|c}
        t & + 1*S \rightarrow -S & -S * B \rightarrow S & -1 * f \rightarrow 0 & 0 \\
        \hline
        T & - 1*S \rightarrow 0 & +S * B \rightarrow -S e^{r(T-t)} & +1 * f \rightarrow K & K- S e^{r(T-t)} > 0
        \end{array}
        + t时刻：套利者按无风险利率r借入现金S，期限为T-t。然后用S购买一单位标的资产，同时卖出一份该资产的远期合约，交割价格为K。
        + T时刻：该套利者将一单位标的资产交割换得现金K，并归还借款本息 $S*e^{r(T-t)}$，实现 $K - S*e^{r(T-t)}$ 的无风险利润

    + 若$K<S*e^{r(T-t)}$ （资产被低估，套利方向是做多标的资产）:

        \begin{array}{c|c|c|c|c}
        t & - 1*S \rightarrow S & +S * B \rightarrow -S & +1 * f \rightarrow 0 & 0 \\
        \hline
        T & + 1*S \rightarrow 0 & -S * B \rightarrow S e^{r(T-t)} & -1 * f \rightarrow -K &  S e^{r(T-t)}-K > 0
        \end{array}
        + t时刻：套利者卖空标的资产获得现金S，以无风险利率r进行投资，期限为T-t。同时买进一份该标的资产的远期合约，交割价为K。
        + T时刻：套利者收到投资本息 $S*e^{r(T-t)}$，并以现金K交割换得一单位标的资产，用于归还卖空时借入的标的资产，从而实现 $S*e^{r(T-t)}-K$ 的无风险利润。
    + 第一列为时间点，第二至第四列为套利策略，第五列为总和现金流

##### 远期价格的期限结构
+ F为T时刻交割的远期；$F^{*}$是$T^{*}$时刻交割的远期
+ 根据定价公式，可得：$F = S e^{r(T-t)}$    $F^{*} = S e^{r^{*}(T^{*}-t)}$    
+ 二者相除消去S，可得：$F^{*} = F e^{r^{*}(T^{*}-t)-r(T-t)}$

##### 标的资产：支付已知红利的资产
+ 支付已知红利资产是指在远期合约到期前会产生完全可预测的现金流的资产，如附息债券和支付已知现金红利的股票等
+ $I = 已知红利的现值$

+ 复制定价：构造投资组合
    + 组合B：
        + 一单位标的资产S & 利率为无风险利率r. 期限为从当前时刻到红利派发日(T-t). 本金为I的负债 *（负债项是为了抵消红利）*
        + 现值：$W_{t} = S_{t} - I$

    + 组合A：
        + 一份远期合约多头f & 现金$K e^{-r(T-t)}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$
    
    + 令二者现值相等，解出f：$f = S_{t} - I - K e^{-r(T-t)}$
    
    + 令$f = 0$，解出K：$K = F = (S_{t} - I) e^{r(T-t)}$

+ 证明（反证法）：
    + 若 $K > (S_{t} - I) e^{r(T-t)}$（资产被低估，套利方向是做多标的资产）

        \begin{array}{c|c|c|c|c}
        t & + 1*S \rightarrow -S & -S * B \rightarrow S & -1 * f \rightarrow 0 & 0 \\
        \hline
        T & - 1*S \rightarrow I * e^{r(T-t)} & -S * B \rightarrow -S e^{r(T-t)} & 1 * f \rightarrow K &  K-(S-I) e^{r(T-t)} > 0
        \end{array}
        + t时刻：套利者按无风险利率r借入现金S，期限为T-t。然后用S购买一单位标的资产，并卖出一份该资产的远期合约，交割价格为K。同时，将T-t期间从标的资产获得的现金收益以无风险利率r贷出至T时刻。
        + T时刻：该套利者将一单位标的资产交割换得现金K，并归还借款本息 $S*e^{r(T-t)}$，同时得到 $I*e^{r(T-t)}$ 的本利收入，最终实现 $K- (S-I)*e^{r(T-t)}$ 的无风险利润

    + 若 $K < (S_{t} - I) e^{r(T-t)}$（资产被高估，套利方向是做空标的资产）
    
        \begin{array}{c|c|c|c|c}
        t & - 1*S \rightarrow S & +S * B \rightarrow -S & +1 * f \rightarrow 0 & 0 \\
        \hline
        T & + 1*S \rightarrow -I * e^{r(T-t)} & -S * B \rightarrow S e^{r(T-t)} & 1 * f \rightarrow -K &  (S-I) e^{r(T-t)}-K > 0
        \end{array}
        + t时刻：套利者卖空标的资产获得现金S，以无风险利率r进行投资，期限为T-t。同时，买进一份该标的资产的远期合约，交割价格为K。
        + T时刻：套利者收到投资本息 $S*e^{r(T-t)}$ ，并以现金K交割换得一单位标的资产用于归还卖空时借入的标的资产，并把该标的资产在T-t期间的现金收益终值 $I* e^{r(T-t)}$ 同时归还标的资产原所有者，最终实现 $ (S-I)*e^{r(T-t)}-K$ 的无风险利润

##### 标的资产：支付已知红利率的资产
+ 支付已知红利率资产，是指在远期合约到期前将产生与该资产现货价格成一定比率的收益的资产，如外汇. 股票指数等，贵金属需要储藏成本，所以存在*负的红利率*或*负的红利*。

+ $q = 已知红利率$ (以连续复利计算)

+ 复制定价：构造投资组合
    + 组合B：
        + $e^{-q(T-t)}$标的资产S，红利收入用于再投资标的资产   （$e^{-q(T-t)}$是为了抵消红利率，最终到期时为 $S$）
        + 现值：$W_{t} = S_{t} e^{-q(T-t)}$

    + 组合A：
        + 一份远期合约多头f & 现金$K e^{-r(T-t)}$
        + 现值：$W_{t} = f + K e^{-r(T-t)}$
    
    + 令二者现值相等，解出f：$f = S_{t} e^{-q(T-t)} - K e^{-r(T-t)}$
    
    + 令$f = 0$，解出K：$K = F = (S_{t}) e^{(r-q)(T-t)}$

+ 证明（反证法）：
    + 若$K >  (S_{t}) e^{(r-q)(T-t)}$（资产被低估，套利方向是做多标的资产）

        \begin{array}{c|c|c|c|c}
        t & + e^{-q(T-t)}*S \rightarrow -S * e^{-q(T-t)}  & -S * e^{-q(T-t)} * B \rightarrow S * e^{-q(T-t)} & -1 * f \rightarrow 0 & 0 \\
        \hline
        T & - e^{-q(T-t)}*S \rightarrow 0 & +S * e^{-q(T-t)} * B \rightarrow -S * e^{(r-q)(T-t)} & +1 * f \rightarrow K & K- S e^{(r-q)(T-t)} > 0
        \end{array}
        + t时刻：套利者按无风险利率r借入现金 $S*e^{-q(T-t)}$，期限为T-t。用来购买 $e^{-q(T-t)}$ 单位标的资产，并卖出一份该资产的远期合约，交割价格为K。同时将T-t期间从标的资产获得的收益都再投资于该标的资产
        + T时刻：该套利者用一单位标的资产交割换得现金K，并归还借款本息 $S*e^{(r-q)(T-t)}$，最终实现 $K- S*e^{(r-q)(T-t)}$ 的无风险利润

    + 若 $K < (S_{t}) e^{(r-q)(T-t)}$（资产被高估，套利方向是做空标的资产）

        \begin{array}{c|c|c|c|c}
        t & - e^{-q(T-t)}*S \rightarrow S & +S * e^{-q(T-t)} * B \rightarrow -S * e^{-q(T-t)} & +1 * f \rightarrow 0 & 0 \\
        \hline
        T & + e^{-q(T-t)}*S \rightarrow 0 & -S * e^{-q(T-t)} * B \rightarrow S * e^{(r-q)(T-t)} & -1 * f \rightarrow -K &  S * e^{(r-q)(T-t)}-K > 0
        \end{array}
        + t时刻：套利者卖空 $e^{-q(T-t)}$ 单位标的资产获得现金 $S*e^{-q(T-t)}$，以无风险利率r进行投资，期限为T-t。同时，买进一份该标的资产的远期合约，交割价格为K。
        + T时刻：由于在T-t期间，标的资产获得收益时套利者都必须通过再卖空标的资产所得收入来偿还，因此套利者所卖空的标的资产数量按连续复利q增长，最后一共卖空了1单位标的资产。T时刻，套利者收到投资本息 $S*e^{(r-q)(T-t)}$，并以现金K交割换得一单位标的资产用于归还卖空时借入的标的资产，最终实现 $S*e^{(r-q)(T-t)}-K $ 的无风险利润

##### 持有成本（Cost of Carry）
+ 持有成本 = 保存成本+利息成本- 标的资产在合约期限内的收益
+ 例子：
    + 不支付红利的股票，持有成本是$c=r$
    + 股票指数，持有成本是$c=r - q$
    + 外币，持有成本是$c=r-r_{f}$
    + 黄金白银等，持有成本$c=r+u$

+ 定价公式可以用持有成本表示：
    + $F = S e^{c(T-t)}$
    + $f = S e^{(c-r)(T-t)} - K e^{r(T-t)}$

##### 不完美市场
1. 存在交易成本的时候，假定每一笔交易的费率为Y
    + 远期价格不再是具体值，而是一个区间：$[S(1-Y)e^{r(T-t)},S(1+Y)e^{r(T-t)}]$
    + 相当于在套利判定的不等式中加入了一个“楔子”

2. 借贷存在利差的时候，如果用$r_{b}$表示借入利率，用$r_{l}$表示借出利率，对非银行的机构和个人，一般是$r_{b}>r_{l}$。
    + 此时的价格区间为：$[Se^{r_{l}(T-t)},Se^{r_{b}(T-t)}]$
    + 借入资金的价格上升，卖空投资的收益下降
    
3. 存在卖空限制的时候，假设由于卖空限制导致的成本比例为X。
    + 此时的价格区间为：$[(1-X)Se^{r(T-t)},Se^{r(T-t)}]$
    + 上限不变，不存在限制；卖空存在X的成本

4. 假设以上情况同时存在：$[(1-X)S(1-Y)e^{r_{l}(T-t)},S(1+Y)e^{r_{b}(T-t)}]$

##### 消费性资产的远期定价

+ 远期价格：$F \leq S e^{c(T-t)}$

+ 原因：
    + 消费性资产具有消费价值（通常称之为便利收益），而远期合约却无法即时消费，消费性资产与其远期之间并不具有完全的可替代性。

    + 因此即使在远期相对价值偏低的时候投资者也不会轻易出售现货，购买远期，从而使得单纯基于风险收益考虑的金融无套利原则不再完全有效。
    

##### $F_{t}$和$S_{t}$. $E(S_{t})$的关系



+ 同一时刻远期（期货）价格与标的资产现货价格的关系：
    + 同一时刻的两者价格高低取决于持有成本（$F_{t} = S_{t} e^{c(T-t)}$）；在远期（期货）到期日，远期（期货）价格将收敛于标的资产的现货价格

    + 标的资产的现货价格对同一时刻的远期（期货）价格起着重要的制约关系（案例3.6 “327”国债期货事件）

    + 远期（期货）与现货的相对价格*只与持有成本c有关*，与未来现货的涨跌预期无关。

    + 价格的领先滞后关系（价格发现功能）

+ 预期收益率：$E(S_{t}) = S e^{y(T-t)}$ （当资产价格服从对数正态分布时成立，y表示预期收益率，$y = r + 风险溢酬$）
    + $F_{t}$和$E(S_{t})$的关系由y和r的大小关系决定

+ 在一个无套利的有效市场中，标的资产和其冗余证券期货之间具有一体化性质，期货的预期收益率总是正好等于标的资产的风险溢酬。

+ 转移风险和管理风险是期货市场的最本质功能。

### 第四章 远期和期货的应用

##### 主要应用：
+ 风险管理
    + 按多/空区分：
        + 多头（买入）风险管理
            + 通过做多远期或期货进行风险管理
            + 适合*担心价格上涨*的投资者，锁定未来买入价格
        + 空头（卖出）风险管理
            + 通过做空远期或期货进行风险管理
            + 适合*担心价格下跌*的投资者，锁定未来卖出价格
    + 按效果区分：
        + 完美的套期保值：
            + 能够完全消除价格风险的套期保值。 （注意：不是指价格不变，而是指未来的价格是确定的）
            + 远期（期货）的到期日. 标的资产和交易金额等条件的设定使得远期（期货）与现货恰好匹配。
        + 不完美的套期保值：
            + 无法完全消除价格风险的套期保值。
            + 常态
        + 原因：
            + 基差风险（Basis Risk）
                + 基差：特定时刻被套期保值的现货价格H 与用以进行套期保值的期货价格G之差（此时现货不一定是期货的标的资产，期货价格不一定等于理论价格）
                + b = H − G
                + 套保到期时基差的不确定性导致了不完美的套期保值
            + 数量风险（Quantity Risk）
                + 可能由于事先无法确知需要套期保值的标的资产规模
                + 可能由于期货合约的标准数量无法完全对冲现货的价格风险
                + 讨论*最优套期保值比率*时，通常不考虑数量风险
            + 相比远期，期货更加标准化，更不易实现完美套期保值
            
    + 运用远期或期货进行套期保值，消除了价格风险，但并不保证盈利。
        + 对于企业来说，完全对冲风险并非是最好的
    + 风险管理应该是针对企业的全局风险进行的，而不是针对局部风险。
        + eg. 炼油企业&原油期货：原油上涨，成品油也会上涨，因此企业应该通过做空来降低整体风险

+ 套利
    + 从第三章的定价方法出发，实际价格偏离理论价格即可套利
    + 在实际操作中，要考虑市场中各种“摩擦”

+ 投机
    + 从本质上看，远期（期货）与其标的资产价格变动的风险源是相同的，只是交割时间不同。
        + 因此远期（期货）与其标的资产现货之间往往存在着良好的替代关系，投机者通过承担价格变动的风险获取收益，既可以通过远期（期货）实现，也可以通过现货实现。

        + 远期（期货）优势在于进入成本低，具有高杠杆效应（双刃剑）。
        + 高杠杆：放大收益/放大亏损
        + $杠杆率 = \frac{1}{保证金比率} $
    + 客观上，套期保值者和套利者往往不足以维持一个市场需要的流动性，此时适度投机可以起到提供流动性的作用。
        + 从这个意义上说，远期（期货）投机者不仅通过承担价格变动的市场风险，也通过承担流动性风险来获取风险收益。

##### 不完美套保
+ 完美的套期保值：
    + 能够完全消除价格风险的套期保值。 （注意：不是指价格不变，而是指未来的价格是确定的）
    + 远期（期货）的到期日. 标的资产和交易金额等条件的设定使得远期（期货）与现货恰好匹配。
+ 不完美的套期保值：
    + 无法完全消除价格风险的套期保值。
    + 常态
+ 原因：
    + 基差风险（Basis Risk）
        + 基差：特定时刻被套期保值的现货价格H 与用以进行套期保值的期货价格G之差（此时现货不一定是期货的标的资产，期货价格不一定等于理论价格）
            + b = H(现货价格) − G(期货价格)
            + 现货多头&期货空头：$ W_{1} = (H_{1}-H_{0})+(G_{1}-G_{0}) = (H_{1}-G_{1})+(H_{0}-G_{0}) = b_{1}-b_{0} $,$b_{1}$是未知的，$b_{0}$是已知的，是否完美套保取决于$b_{1}$
            + 分解$b_{1}$：$b_{1}= H_{1}-G_{1} = (H_{1}-S_{1})+(S_{1}-G_{1})$,$S_{1}$为标的资产价格，若$H_{1}=S_{1}=G_{1}$,则$b_{1}=0$，达成完美套保
        + 套保到期时基差的不确定性导致了不完美的套期保值
        + 由于$b_{1}$的变动往往小于G的变化，所以不完美套保往往也能对冲风险
        +  $b_{0}-b_{1}$减小→多头套保有利；$b_{1}-b_{0}$增大→空头套保有利
    + 数量风险（Quantity Risk）
        + 可能由于事先无法确知需要套期保值的标的资产规模
        + 可能由于期货合约的标准数量无法完全对冲现货的价格风险
        + 讨论*最优套期保值比率*时，通常不考虑数量风险
        + 相比远期，期货更加标准化，更不易实现完美套期保值

##### 套保策略
+ 合约种类
    + 足够流动性 & 与现货资产高度相关 
+ 到期日
    + 对于实物交割的期货而言，要避免在*期货到期的月份*中持有期货头寸，以防止逼仓。（避免麻烦的交割流程 & 交割月频繁的价格波动）
    + 在到期时间无法完全吻合时，通常选择比所需的套期保值*月份略晚但尽量接近*的期货品种。（避免期货价与现货价差距过大）
    + 所需套期保值时间较长时，可使用套期保值*展期*（快要到期时，平仓原有合约，购买期限更靠后的合约），但可能给套期保值者带来额外的风险。
+ 头寸方向
    + 价格上涨时不利→多头套保
    + 价格下跌时不利→空头套保
+ 交易数量

##### 套保比率
+ 套期保值比率（Hedge Ratio）是指用于套期保值的资产（如远期和期货）头寸对被套期保值的资产头寸的比率
+ $套期保值比率 = \frac{用于套保的资产头寸}{被套保的资产头寸}$
+ 最优套期保值比率：能够最有效. 最大程度地消除被保值对象价格变动风险的套期保值比率。（最小的套保比率）
    + 存在基差风险时，最优套期保值比率很可能不为1

##### 最优套保比率（风险最小化的套保比率）

设1单位现货，n单位期货多头套保：

套保组合：$\Delta \Pi = n\Delta G- \Delta H$

最优套保条件(套保组合的价格变化不依赖于现货价格)：$\frac{\partial \Delta \Pi}{\partial \Delta H} = \frac{\partial (n\Delta G- \Delta H)}{\partial \Delta H} = n \frac{\partial \Delta G}{\partial \Delta H} -1 = 0$ → $n = \frac{\partial \Delta H}{\partial \Delta G}$

由于收益率$r_{H} = \frac{H_{1}-H_{0}}{H_{0}} = \frac{\Delta H}{H_{0}},r_{G} = \frac{\Delta G}{G_{0}}$,

所以$n = \frac{\partial \Delta H}{\partial \Delta G} = \frac{H_{0}* \partial r_{H}}{G_{0}* \partial r_{G}}$

上述最优套期保值比率n仅针对单位价值变动，因此，实际的最优套期保值数量N还应在n的基础上考虑具体的头寸规模,此时现货头寸为N

$N = n*\frac{Q_{H}}{Q_{G}} = \frac{\partial (\Delta H)*Q_{H}}{\partial (\Delta G)*Q_{G}} = \frac{Q_{H}*H_{0}* \partial r_{H}}{Q_{G}*G_{0}* \partial r_{G}}$

##### 最小方差套保比率
此时用方差$\sigma_{\Pi}^{2}$表示组合的风险

$\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$

对n求导，令$\frac{\partial (\sigma_{\Pi}^{2})}{\partial n} = 2\sigma_{G}^{2} n -2 \sigma_{HG}^{2} n =0$

由于$\frac{\partial^{2} (\sigma_{\Pi}^{2})}{\partial n^{2}} = 2\sigma_{G}^{2}>0$恒成立，所以最小化$\sigma_{\Pi}^{2}$的n为$n = \frac{\sigma_{HG}^{2}}{\sigma_{G}^{2}}$

##### 最小方差套保比率& OLS

注意到，若对$\Delta H = \beta \Delta G + \alpha + \epsilon$进行OLS线性回归，则有$\beta = \frac{Cov(H,G)}{Var(G)}$

因此，实践中可以通过OLS的$\hat{\beta} = \frac{\Sigma x_{i}y_{i} - N \bar{x}\bar{y}}{\Sigma x_{i}^{2} -N \bar{x}^{2}}$来估计$n$

ΔH 与ΔG 的期间应与实际套期保值期长度相同，且时期之间不宜重合（Overlapping）。但实际应用中ΔH与ΔG多用日度数据，与套保长度不符。

##### 基于收益率的OLS
不用价格，而用收益率进行回归：$r_{H} = b r_{G} + a + \epsilon$

此时套保比率为：$n = \hat{\beta} = \frac{Cov(\Delta H,\Delta G)}{Var(\Delta G)} = \frac{Cov(\Delta H_{0}r_{H},\Delta G_{0}r_{G})}{Var(\Delta G_{0}r_{G})} = \hat{b}\frac{H_{0}}{G_{0}} $

实际套保数量则是：$N = n \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{H_{0}}{G_{0}} \frac{Q_{H}}{Q_{G}} = \hat{b}\frac{V_{H}}{V_{G}} $

##### 套保有效性
最小方差套保比率的有效性可以通过检验风险降低的百分比来确定。因此，定义套期保值有效性为：$e^{*} = \frac{\sigma_{H}^{2}-\sigma_{\Pi}^{2}}{\sigma_{H}^{2}} $

将 $\sigma_{\Pi}^{2} = \sigma_{H}^{2} + n^{2}\sigma_{G}^{2} - 2n \sigma_{HG}^{2}$ 代入，化简可得 $e^{*} = \rho_{HG}^{2} = \rho_{r_{H}r_{G}}^{2} = 线性回归(r_{H} = b r_{G} + a + \epsilon) 的判别系数R^{2} $

显然，$e^{*}$越接近1，套保效果越好

+ 实际上，只要是基于同一或是类似标的资产的现货与衍生产品之间，都可以相互进行风险对冲。
    + 例如期货可以用来为相同或类似标的资产的期权空头进行风险对冲。

+ 注：静态套保只适用于线性衍生品。对于债券. 期权等非线性（payoff不再只是直线）衍生品，只能用动态套保。在动态套保时不能运用统计方法（线性回归）来求n, 而只能通过理论关系求偏导。


### 第五章 多种期货与远期（期中考到这儿）

##### 股指期货
+  标的资产：股价指数
    +  运用统计学中的指数方法编制而成的. 反映股市中总体股价或某类股票价格变动和走势情况的一种相对指标。

+  股指期货
    +  以股价指数作为标的资产的股价指数期货，交易双方约定在将来某一特定时间交收“一定点数的股价指数”的标准化期货合约。

+  特殊性质
    +  现金结算而非实物交割
    +  合约规模非固定
        + $合约规模 = 股指期货的价格点数 × 每个指数点所代表的金额（固定的合约乘数）$

+ 一般定价公式：$F = S e^{(r-q)(T-t)} $ ,q表示红利率

+ 指数套利（Index Arbitrage）
    + “程序交易”（Program Trading）：每次买卖都需要买卖一揽子股票，交易所一般借助程序来进行交易
    + 1987年10月19日“黑色星期一”
        +当时 纽约股票交易所的交易系统超负荷运行，无法处理大量交易指令，指数套利无法进行。 

+ 套期保值
    + 管理系统性风险
    + 多为交叉套期保值


##### 股指期货的最小方差套保比率
+ CAPM：$R_{i} = \beta_{i}(R_{M} - r) + r $，其中 $\beta_{i} = \frac{Cov(R_{i},R_{M})}{Var(R_{M})}$
    →类比：$  \hat{b} = \frac{Cov(\Delta r_{H},\Delta r_{G})}{Var(\Delta r_{G})}$

    + 注意：CAPM中不涉及期货价格，需要用*未来的*$\hat{\beta}$才能近似$\hat{b}$
    + 此时 $n = \hat{b} \frac{H_{0}}{G_{0}} = \beta \frac{H_{0}}{G_{0}}$
+ 假定：
    1. 被套期保值的股票组合与市场指数$R_{M}$ 之间的β系数等于股票组合与股指期货之间的β系数
    2. 所采用的β系数等于套期保值期间真实的β系数
    + 满足以上假定时，β系数是股指期货最小方差套期保值比率n的一个良好近似

+ 股票头寸↔短期国库券头寸(基于CAPM的复制定价)：
    + $+股票-期货 = +债券$ → $+债券+期货 = +股票$
    + 构造短期国库券多头等价于将系统性风险降为零


##### 调整而非消除系统性风险
+ 可以利用股指期货，根据自身的预期和特定的需求改变股票投资组合的beta系数（由β 变为β*），从而调整股票组合的系统性风险与预期收益
+ 此时套保比率为 $n = (\beta^{*} - \beta)$
+ 套保数量为 $N = (\beta^{*} - \beta) \frac{V_{H}}{V_{G}}$

#### 外汇远期

+ 外汇远期的持有成本为$r_{f}$,即持有外汇时获得的外国银行无风险利率
##### 远期价值&远期汇率

1. 远期价值：$f = S e^{(r_{f})(T-t)} - K e^{r(T-t)} $
2. 远期汇率（价格）：$F = S e^{(r-r_{f})(T-t)} $
+ 利率平价关系：
    + 若 $r_{f} > r$ , 外汇远期贴水 $F_{t}<S_{t} $；
    + 若 $r_{f} < r$ , 外汇远期升水 $F_{t}>S_{t} $。
    + 注：远期升贴水是指远期汇率与当前即期汇率的相对高低，并不意味着外汇真实的升值与贬值

#### 远期利率
+ 远期利率是指现在时刻 $(t+t^{*}时刻)$的将来一定期限 $(T+t^{* }时刻)$ 的利率。 
    + $([t+t^{*},T+t^{*}])$，$(t+t^{*})*(T+t^{*})$表示 $t+t^{*} $时刻起 $(T-t)$期间的利率
+ 即期利率是指当前时刻(t时刻)起一定期限(T时刻)的利率。 $([t,T])$

设利率期限结构如下：

![](./interest.png)

为了使$T^{*}$时刻终值相等，可得：$$e^{[r(T-t)+r_{F}(T^{*}-T)]} = e^{r^{*}(T^{*}-t)}$$
因此：$$r(T-t)+r_{F}(T^{*}-T) = r^{*}(T^{*}-t)$$（对比：远期价格的期限结构）
$$r_{F} = \frac{[r^{*}(T^{*}-t) - r(T-t)]}{(T^{*}-T)}$$
化简可得：$$r_{F} = (r^{*}-r)\frac{T-t}{T^{*}-T} +r^{*} $$
+ 如果即期利率期限结构在 $T^{*}-T$期间是向上倾斜的，即$r^{*}>r$，则 $r_{F}>r^{*}$（升水）； 反之，则$r_{F}<r^{*}$（贴水）。


##### 远期利率协议（FRA）
+ 远期利率协议（FRA）是买卖双方同意从未来某一商定的时刻开始的一定时期内按*协议利率*借贷一笔数额确定. 以具体货币表示的*名义本金*的协议。
    + 在t时刻与银行约定，在T时刻向银行以$r_{F}$利率借款/存款，在$T^{*}$时刻还款/提款

+ FRA的特征
    +  在T时刻进行现金结算，
        + 结算金额为 $T^{*}$时刻利差贴现至 $T$时刻的贴现值
        + 贴现率为真实利率 $r$
        + 由亏损方支付给盈利方
        + 现金结算：只交割差价，不实际交割贷款

    +  名义本金：合约中的本金只是名义本金，并不一定涉及贷款的实际交付

    +  头寸：Long/Short
        + Long: Fixed-rate payer (固定利率支付者即为FRA的多头)

    +  报价：3×9（3月后，借6个月） SHIBOR（约定的市场参考利率） 2.86（合同的约定利率）
    
+ 注：
    + FRA的买方是名义借款人，其订立FRA的目的主要是为了规避利率上升的风险。
    + FRA的卖方是名义贷款人，其订立FRA的目的主要是为了规避利率下降的风险。

##### FRA定价
1. t时刻存在2个期限相同$(T,T^{*})$的FRA，一个FRA的协议利率为市场协议利率$r_{F}$，一个为$r_{K}$
2. 由于第一个FRA协议利率为$r_{F}$，故$f=0$
3. 第二个FRA的价值即为二者*支付利息*差额的贴现值：$f = e^{-r^{*}(T^{*}-t)}(A e^{r_{F}(T^{*}-T) - r_{K}(T^{*}-T)}) $

#### 利率期货

+ 利率期货：是指以*利率或利率敏感性证券（如债券）*作为标的资产的期货合约。
    + 利率：CME的欧洲美元期货
        + 标的资产为*自期货到期日起3个月*的欧洲美元（存放于美国境外的美元）定期存款。
    + 债券：中金所的2年期. 5 年期. 10年期国债期货
        + 这些“期限”对应的是交割时的债券期限，

+ 欧洲美元期货合约
    + 合约规模 
        + 2 500 美元 x合约 IMM 指数
    + 报价
        + $合约 IMM 指数 = 100-R = 100-100*r $，r为报价中隐含的期货利率
            + 期货利率含义与远期利率类似
            + IMM指数的变动=期货利率的变动×100，方向相反
                + 担心利率上涨者，看多利率，应当*做空*欧洲美元期货

        + $R= 合约到期月份第三个周三进行现货结算的 3 个月期美元 LIBOR 水平$

        + 期货利率1个基点（0.01%）的变化对应着期货价格0.01的变化，进一步对应着每份合约盈亏25美元。
            + $2500美元×合约IMM指数=2500美元×(100-R)=2500美元×100 ×(1-r)$
        + 盈亏计算过程中不考虑贴现
        + $ 1 个利率基点 = 0.01% = 0.01 价格点 = 25 美元/合约$

    + 交易月份

    + 最后交易日与结算日
        + 到期月第三个周三往回数的第二个伦敦银行工作日伦敦时间上午 11:00(芝加哥时间上午 5:00)。
        
    + 结算方式
        + 根据到期结算日伦敦时间上午11 ： 45 公布的 3 个月期美元 LIBOR 进行现金结算。 
        + 最后结算价将四舍五人至小数点后 4 位， 即 0.000 1% ,对应着每份合约 0.25 美元。
        

##### 期货利率&远期利率
+  欧洲美元期货合约与远期利率协议都锁定了未来一定期限的利率。
+  1 年以下的到期期限，期货利率≈ 远期利率

+  长期：差异不能忽略
    +  一次性到期/每日盯市结算和保证金：远期利率较低（每日盯市可以不采用自融资策略，更加灵活）
    +  盈亏结算时贴现/无贴现：远期利率较低


1. 远期利率协议报出的是远期利率，而利率期货报出的通常并非期货利率，而是与期货利率反向变动的特定价格，期货利率隐含在报价中。
2. 由于上述区别，利率期货结算金额为T时刻协议价与结算价之差，远期利率的结算金额则为T*时刻利差的贴现值。
3. 利率期货存在每日盯市结算与保证金要求，加上结算金额计算方式的不同，决定了远期利率与期货利率的差异。
4. 在远期利率协议中的多头是规避利率上升风险的一方，而利率期货的多头则是规避期货价格上升风险，即规避利率下跌风险的一方。
5. 远期利率协议通常采用现金结算，而利率期货可能需要实物交割，期货交易所通常规定多种符合标准的不同证券均可用以交割，使得利率期货相对复杂

##### 国债现货的全价&净价
+ 报价时通常报出面值每100元的价格
+ $全价=净价+应计利息$,国债现货使用净价而非全价来报价
    + 应计利息：上一个付息日以来的利息（按比例计算），是债券的前持有人应得的利息       
    + $ 100元面值现货的应计利息 = \frac{票面利率*100}{每年付息次数} * \frac{现货交割日(t)-上一个付息日}{当前付息周期实际天数} $
    + （实际天数比实际天数）
+ 付息日日期即为到期日日期
+ 付息周期为 $[上一个付息日,下一个付息日)$，算头不算尾
+ 定价公式里面使用的资产价格都是全价

##### 可交割券与标准券
+  可交割券：期货空方可以选择发行期限不高于7年，合约*到期月份首日*剩余期限为4-5.25年的任何记账式附息国债均为该期货合约的可交割券。
    +  主要特征：不唯一. 由交易所确定并公布. 数量只增不减（财政部还会发新债）
    +  一种现券可以是多种期货的可交割券，一个期货也可以有多种可交割券
    
+  标准券：面值为100万，息票率为 3% ，在交割月首日的剩余到期期限为5年整的虚拟债券（作为比较标准，并不存在于市场交易），是其他实际可交割债券价值的衡量标准。
    +  主要特征：虚拟国债. 整数期限. 票面利率3%. 面值100万
    +  （以3%贴现时，为平价债券），在交割月首日，不存在应计利息，标准券的净价=全价=100


##### 转换因子
+  转换因子：$转换因子 = \frac{可交割券j的净价}{标准券净价} = \frac{可交割券j的净价}{100} $
    + （因为标准券为平价债券且不含应计利息）
+  中金所规定转换因子计算：每1元面值的可交割债券的未来现金流，按3%的年到期收益率贴现到交割月首日的价值，再扣掉该债券每1元面值应计利息后的余额。
    +  时间调整（只算月份之差）
    +  本质上是净价（贴现现金流-应计利息）
    +  交易所公布
+  转换因子的具体计算不作要求（课外阅读）
    
+ Note: 
    1. 同一个债券对于不同国债期货合约的转换因子是不同的；
    2. 给定国债期货的可交割券的转换因子不变。


##### 国债期货的净价和全价
1. 国债期货使用标准券期货净价报价，因此市场上看到的国债期货行情价格就是*标准券*的期货净价

2. $可交割券的期货净价_{j} = 标准券期货净价*转换因子_{j}$

3. $国债期货空方交割时收到的现金 = 期货交割全价 = 可交割券的期货全价_{j} = 可交割券的期货净价_{j} + 期货应计利息_{j} = 标准券期货净价*转换因子_{j} + \frac{可交割券票面利率*100}{每年付息次数}*\frac{配对缴款日(T+2)-上一个付息日}{付息周期实际天数}$

    + 其中$ 期货应计利息 = \frac{可交割券票面利率*100}{每年付息次数} * \frac{配对缴款日(T+2)-上一个付息日}{付息周期实际天数} $
    + 此时的T(交割意向申报日/交割日)只能是交易日（周末不交易）
    
    + T+1：卖方交券日，T+2：配对缴款日，T+3：买方收券日

##### 国债现货&国债期货
+ 国债现货 
    + 每天的应计利息都会变化 

    + 现货的净价仅用于报价

    + 现货交割和计算盈亏时使用的都是全价

+ 国债期货
    + 一旦确定具体标的债券和配对缴款日，应计利息不变
 
    + 在期货交割之前，期货交易报价. 计算期货盈亏. 盯市结算中使用的都是期货净价

    + 应计利息和全价只有在期货交割. 为期货定价等时候用到

##### 国债期货的交割全价
+ 国债期货交割时空方应收到（多方应付出）的现金，即为国债期货的交割全价。
+ 每交割100元面值的特定债券应收到的现金为：
    + $期货交割结算价×交割券转换因子+交割券配对缴款日应计利息$

##### 转换因子的不足
+ 理想的转换因子应该是各债券在交割时刻的真实价格之比
+ 现实中转换因子的不足：
    1. 由于事先无法预知可交割券在交割月的真实到期收益率，计算转换因子时所有债券使用了同一个贴现率3%，这将导致转换因子出现误差；
    2. 在计算转换因子时，由于无法预知真实交割时刻，中金所规定使用整数月份，这进一步使得转换因子成为一个近似值；
    3. 在计算转换因子时，中金所对一年支付一次利息和一年支付两次利息的债券都使用3%的贴现率，实际上应对计息频率进行转换才能使两个贴现率等价。
+ 由于转换因子诸多假设，天然导致不同债券之间并非完美公平转换——相对合算和不合算
+  Question:作为有选择权的卖方，会选哪个券交割？

##### 最合算交割券（CTD券，cheapiest to deliver）
+ 最合算交割券：购买交割券所付的价格与交割期货时空方收到的现金之差最小的债券。
+ 目标：交割日的交割成本最小
    + $ 交割成本_{j} = 现券全价_{j} - 期货全价_{j} = 现券报价_{j} -期货结算价*交割券转换因子_{j} $
    + （T时刻，现券和期货的应计利息基本相等）
+ 准CTD券
    + 因为只有到期货交割时我们才能获知交割时的现券报价和期货结算价。这意味着在国债期货到期交割前，哪个券是真正的CTD券是无法预知的。
    + 所以只能退而求其次，根据当时的信息条件判断哪个券未来交割时最可能成为CTD券，即在国债期货到期交割前，我们只能确定“准CTD券”。市场情况变化，准CTD券也可能发生变化。
    + 判断方法：IRR方法 & 经验方法

##### 国债期货定价
+ 假设：
    1. 不存在择券期权
    2. 不存在择时期权
        + 因为这两个权利属于空方，所以理论定价会高于实际价格
    3. 准CTD券和交割日期已知

+ 定价公式：
    + 由于准CTD券和交割日期已知，所以可得 $S_{t} = 现货净价 + 应计利息 $
    + 所以可得期货全价： $F_{t} = (S_{t}-I_{t}) e^{r(T-t)} $
        + $I = t~T时间段的付息贴现至t时刻$
    + 已知：$期货全价^{j} = 期货净价 * 转换因子^{j}+ 期货应计利息^{j} $
        + 所以 $期货净价 = \frac{期货全价^{j}- 期货应计利息^{j}}{转换因子^{j}} = \frac{F_{t}^{j}- 期货应计利息^{j}}{转换因子^{j}}$
        + 期货应计利息：前一个付息日至T+2时刻的应计利息

##### 计算国债期货的报价
+ 已知：准CTD券的现货报价（净价）以及转换因子. 预期的交割日期
+ t:定价日, T:交割日, T+2:配对缴款日
1. 准CTD券的现货净价 → 现货应计利息 → 准CTD券的现货全价
    + $ 现货应计利息 = \frac{100*票面利率}{付息次数}*\frac{定价日-上一个付息日}{下一个付息日-上一个付息日} $

2. 预期的交割日期 → 合约期限内的现金红利 → $ F = S e^{r(T-t)} $ → 准CTD券的期货全价
    + $ I = \Sigma \frac{CF_{i}}{(1+r)^{i}} $
    + 期限为$[t,T)$

3. 预期的交割日期 → 期货应计利息 → 准CTD券的期货净价
    + $ 期货应计利息 = \frac{100*票面利率}{付息次数}*\frac{配对缴款日-上一个付息日}{下一个付息日-上一个付息日} $

4. 准CTD券的转换因子 → 标准券期货净价（国债期货报价）


##### 国债期货定价存在的问题
1. 由于国债期货卖方拥有择券期权和择时期权，本章所定出的国债期货价格通常会高于真实的国债期货价格。

2. 国债期货只能施行正向套利策略：
    + 对于国债期货来说，只有在国债期货市场价格过高时，实施“买现货卖期货”的套利策略才可能盈利；
    + 但在国债期货市场价格过低时，投资者并不能像普通期货那样进行“买期货卖现货”的套利交易，因为*未来期货交割时间和交割券种的选择权*都在期货空方手中，套利者无法确定赚取无风险利润。

3. 国债期货价格经常小于债券现货价格
    + 期货存续期内现货可能支付票息
    + 现券票息经常高于短期利率
    + 国债期货空方有隐含期权
    + $国债期货价值 = 期货合约价值-择券期权价值-择时期权价值$

4. 对于每一个可交割券，其对应的国债期货基差定义为：
    + 用净价定义：
    + $ 国债期货基差_{j} = 可交割券净价_{j} - 国债期货报价 * 转换因子_{j}$

#### 利率风险管理

##### 资产价值的利率风险
+ 资产的利率风险一般被表述为资产价格变动的百分比对到期收益率变动的敏感性。
+ 由于资产价值P是到期收益率y的函数, $P = f(y)$，资产价值的利率风险可以用泰勒展开表示如下：
    + $ \frac{d P}{P} = \frac{1}{P} * \frac{dP}{dy} + \frac{1}{P*2!} * \frac{d^{2}P}{d y^{2}} +...$
    + 其中第一项为久期(duration)，第二期为凸性

##### 久期
+ 久期(Duration)：资产价格变动的百分比对到期收益率变动的一阶敏感性
    + $D = - \frac{d P/P}{dy} = -\frac{1}{P} \frac{d P}{dy} $

+ 久期*一般*为正，此时利率上升，债券价格下降

+ 久期反映了资产价格利率风险的*主要*部分，并非全部，还存在二阶项和高阶项

+ 久期越大，资产的利率风险越大；反之则越小
    + 久期越大，债券价格对利率变动越敏感，利率风险越大

+ 如果资产价值是利率的*非线性函数*，显然其久期并非常数，而是会*动态变化*的

##### 货币久期（美元久期）
+ 货币久期(Dollar Duration)：到期收益率变动引起的价值变动金额：
    + $\$D = D * P= \frac{dP}{dy} $
+  对于非线性产品而言，货币久期也是*时变*的。
+  对衍生品而言，货币久期尤为重要。
    + 因为很多衍生品刚签约时的*合约价值为零*，无法计算合约的久期，只能计算合约的货币久期。
+  对于期货合约来说，由于期货合约价值的变动源于期货价格的变动，因此如果不考虑合约规模的影响，期货合约的货币久期就等于期货价格的货币久期。

##### 麦考利久期
+ 假定债券价格为现金流贴现值： 
    + $P = \Sigma_{i=1}^{n} \frac{CF_{i}}{(1+y)^{i}} +\frac{A}{(1+y)^{n}} $
+ 计算久期D: $D = -\frac{1}{P} \frac{dP}{dy} = \frac{1}{1+y} *[\Sigma \frac{CF_{i}}{(1+y)^{i}} + \frac{A}{(1+y)^{n}}]*\frac{1}{P}  $
    + 其中：$[\Sigma \frac{CF_{i}}{(1+y)^{i}} + \frac{A}{(1+y)^{n}}]*\frac{1}{P} $ 即为麦考利久期
        + 意义：按现金流加权的平均到期期限
    + 真正衡量风险的久期为 $ D$，被称为修正久期

+ 常见误解：
    + ×：麦考利久期就是不含权债券的久期 
        + √：修正久期才是不含权债券的久期
    + ×：久期一定是期限的加权 
        + √：久期与期限加权之间并没有必然联系
    + ×：久期的单位一定是年 
        + √：久期的单位并不必然为年

+ $ D = -\frac{1}{P} \frac{dP}{dy}$才是最具一般性. 最能反映久期性质的定义与公式
+ 连续复利情况下的定价：
    + $P_{t} = \Sigma{CF_{i}}{e^{-y(t_{i}-t)}}$
    + 对其求导，可得久期 $ D_{t} = -\frac{1}{P} \frac{dP}{dy} = -\frac{\Sigma{CF_{i}}{e^{-y(t_{i}-t)}}}{P}  (t_{i}-t)$
    + 此时麦考利久期等于修正久期
    + 在使用连续复利到期收益率时，普通债券的久期就是现金流的加权平均期限，无需再调整

##### 有效久期
+ 用于不存在解析解的定价模式
+ $D = \frac{P_{-}-P_{+}}{2*P*\Delta y} $
    + $P_{-} $与$P_{+} $分别代表到期收益率下跌和上升时所达到的资产价格
    + 核心方法：中心差分

##### 基点价格值
+ 基点价格值(DV01或BPV）
    +  利率变化1个基点引起的资产价格变动
    +  与1个基点的货币久期十分接近
+ $ BPV = \frac{P_{-0.01\%}-P_{+0.01\%}}{2} $
    + 用差分取代微分

+ 利率远期和利率期货的利率敏感性估计取决于其标的资产的利率敏感性和远期（期货）本身价值变化的计算方式。
    + 例如：欧洲美元期货通常使用基点价格值这一利率敏感性指标，即1个基点对应着一份欧洲美元期货合约的价格变动为25美元。

##### 国债期货的久期

+ 回忆：国债期货的定价
    + $G = \frac{期货全价_{准CTD} - 期货应计利息_{准CTD}}{转换因子_{准CTD}}$
    + $期货全价_{准CTD}F = (S-I)e^{r(T-t)} $

1. 准CTD券期货价格F的久期和货币久期
    + 假设：到期收益率y通过影响现货价格S来影响F，y与利率r无关
    1. 已知$ F= (S-I) e^{r(T-t)} $
    2. 货币久期:$$ \$D_{F} = -\frac{dF}{dy} = -e^{r(T-t)}\frac{dS}{dy} = e^{r(T-t)}\$D_{S} $$
    3. 久期:$$ D_{F} = -\frac{1}{F} \frac{dF}{dy} = -\frac{e^{r(T-t)}}{(S-I) e^{r(T-t)}} \frac{dS}{dy} = -\frac{1}{(S-I)} \frac{dS}{dy} \approx -\frac{1}{S} \frac{dS}{dy} = D_{S} $$

2. 国债期货报价G的久期和货币久期
    + 久期：
    $$ D_{G} = -\frac{1}{G}\frac{\delta G}{\delta y} = -\frac{1}{\frac{F -应计利息}{转换因子}}*\frac{\delta ({\frac{F -应计利息}{转换因子}})}{\delta y} = -\frac{1}{F -应计利息} * \frac{\delta F}{\delta y} $$
    $$D_{G} = -\frac{1}{F -应计利息} * D_{F}*F \approx -\frac{1}{F} * D_{F}*F = D_{F} = D_{S} $$
    + 此时G的久期≈准CTD券的久期
    + 货币久期：
    $$ \$D_{G} = -\frac{\delta G}{\delta y} =\frac{\delta ({\frac{F -应计利息}{转换因子}})}{\delta y} = -\frac{1}{转换因子}\frac{\delta F}{\delta y} = \frac{1}{转换因子}\$D_{F} \approx \frac{1}{转换因子}\$D_{S} $$
    + 在现实中运用国债期货进行利率风险管理时，我们是在用期货合约价值的变动去对冲其他资产的利率风险，而每天期货合约价值的变动是由标准券期货报价的变动引起的。因此需要计算国债期货报价的久期和货币久期
+ 注意：
    1. 由于债券和国债期货价格都是利率的非线性函数，其久期和货币久期都是动态变化的；
    2. 国债期货报价的久期和货币久期都是基于准CTD券的利率敏感性指标计算得到的。准CTD券一旦变化，期货的利率敏感性也会随之变化。

##### 基于久期的利率风险管理

+ 由于利率敏感性资产价格与利率存在非线性关系，因此无法进行静态套保，只能进行动态套保。
+ 最优的利率风险套期保值比率n是使得*整个套保组合价值*（而不是其变动率）对利率的一阶敏感性为零的套期保值比率: $\frac{d\Pi}{dy}=0$
    + (目标： $arg_{min}\frac{d\Pi}{dy}$)
+ n实际上是使得套期保值组合的*货币久期*或基点价格值（BPV）为零的套期保值比率

以现货多头和期货空头的空头套期保值组合为例:
+ $d\Pi = dH-n*dG $

+ 令 $\frac{d\Pi}{dy} =0$,可得 $n = \frac{\frac{dH}{dy}}{\frac{dG}{dy}} = \frac{D_{H}*H}{D_{G}*G}$
+ 可得最优套保数量N：$N = \frac{Q_{H}}{Q_{G}}*n = \frac{D_{H}*V_{H}}{D_{G}*V_{G}}$


有时候，并不需要将久期 $D_{H}$对冲至0，只需要将其控制于一个基准 $D_{H}^{*}$

+ 设投资组合的原久期为 $D_{H}$，目标久期为 $D_{H}^{*}$，则需要交易的利率敏感性证券的份数N为
    + $ N = \frac{D_{H}^{*}-D_{H}}{D_{G}}*\frac{V_{H}}{V_{G}}$
    + $V_{G}$ 是一份期货合约按标准券报价计算的合约规模
    + 上式为负时，说明需要进行反向操作，即原来是多头的要卖出，原来是空头的要买进


##### 久期的局限性

+ 久期有着天然的局限性：
    + 久期仅仅是资产价格对利率的一阶敏感性，无法反映和管理资产价格的全部利率风险，当利率变化较大时这个缺陷尤其显著；
        + 有时候需要考虑凸性
    + 久期的定义建立在利率曲线发生平移，即所有期限的利率变化幅度相等的假设基础之上，这是一个不符合现实的假设
        + 没有考虑不同期限的收益率曲线形状不同的情况

### 第六章 互换

+ 互换（Swaps）是两个或两个以上当事人按照商定条件，在约定的时间内交换一系列现金流的合约。
+ 互换可以看作是一系列远期的组合。
+ 最重要和最常见的互换：
    + 利率互换（ Interest Rate Swap，IRS）
    + 货币互换（ Currency Swap）
    + 总收益互换（Total Return Swap，TRS）
    + 信用违约互换（Credit Default Swap，CDS）

##### 利率互换

+ 利率互换，是指双方同意在未来的一定期限内根据同种货币的*相同名义本金*交换现金流的一种*场外交易*的金融合约，其中一方的现金流根据事先选定的某一*浮动利率*计算，而另一方的现金流则根据*固定利率*计算。
+ 常见期限包括1年. 2年. 3年. 4年. 5年. 7年与10年，30年与50年的互换也时有发生。
+ 常见的互换频率包括季度. 半年和年度

案例：国内第一笔利率互换

+ 协议时间：2007年1月18日
+ 交易双方：兴业银行与花期银行
+ 利率互换期限：自2007.1.18起期限为1年
+ 名义本金：未披露（假设为1元）
+ 结算方式：现金结算，本金不需要交换（降低风险敞口）
+ 约定：每三个月交换一次现金流如下：
    + $ 兴业银行→^{\frac{2.808\%}{4}}花旗银行$
    + $ 兴业银行^{\frac{3月期SHIBOR}{4}}←花旗银行$

#### 货币互换
+ 定义
    + 货币互换，是在未来约定期限内将一种货币的本金和固定利息与另一货币的等价本金和固定利息进行交换
    + $货币互换(currency \space swaps) \neq 外汇掉期(foreign \space exchange \space swaps)$
        + 外汇掉期锁定了相对汇率
+ 利率互换与货币互换之差异：
    + 利率互换：通常无需交换本金，只需定期交换利息差额；
    + 货币互换：期初和期末须按照约定的汇率交换不同货币的*本金*（因为名义本金的币种不同），期间还需定期交换不同货币的利息。

+ 流程：
    + ![货币互换](currency_swap.jpg)
    + 假设汇率不变，此时雷公司的实际借款成本为$欧元利率+（支付的美元利率-收到的美元利率） = 4.35%+(6.5%-6.1%) = 4.75% $
    + 由于其中有一部分美元利息，所以雷公司仍要承担汇率风险

#### 总收益互换 TRS

+ 总收益互换：Total Return Swap
    + 在未来约定期限内将一种或一篮子资产的总收益（包括现金流与资本利得或损失）与等值浮动利率债券的利息加或减价差进行交换。
    + $ 支付方 {}^{资产总收益率}\rightleftarrows_{浮动利率+一定的净息差} 接收方 $

+ 特点：
    + 总收益互换的一个重要特征是合约双方并不转让资产的实际所有权，只是转让*收益权*。
        + 对于支付方来说，相当于已经把资产卖掉了，但是不会体现在资产负债表上
        
    + 总收益互换由总收益端（称为总收益支付方，或者总收益互换*卖方*）和融资端（称为总收益接收方，或者总收益互换*买方*）组成。
    + 总收益端的资产可以是单一股票. 指数. 公司债. 抵押贷款. 不动产. 大宗商品. 外汇. 贵金属等。
        + 范围很广

    + 总收益互换优点：

        + 节约交易成本

            + 可以规避买卖资产带来的价格风险和交易费用

        + 提高杠杆

            + 可以实现融资融券的效果

                + 融资（借钱买股）：一边当买方，一边当卖方

                + 融券（借股卖钱）：用TRS卖出手上的股票

        + 规避监管

            + 方便在全球配置资产
            
        + 注意：需用保证金控制风险

#### 信用违约互换 CDS
+ 信用违约互换：Credit Default Swap
    + CDS的买方以一定的费用为代价将基础资产的信用风险转移给CDS的卖方，相当于购入了一份信用保险
    + $ 基础资产 \rightarrow CDS买方 \space \space {}^{支付保费}\rightleftarrows_{发生信用事件时进行赔付} CDS卖方 $
    + 相当于担保信用的保单
+ The Big Short

#### 其它互换
1. 交叉货币利率互换
    + 一种货币的固定利率换另一种货币的浮动利率
2. 基点互换
    +  Constant Maturity Swap (CMS)：LIBOR与特定期限的互换利率
    +  Constant Maturity Treasury Swap (CMT): LIBOR与特定期限的美国国债利率

7. 后期确定互换：浮动利率在期末确定

    + 交换两种货币的浮动利率，但按照一种货币的相同名义本金计算，从而可以利用不同市场的利率差异，而不用考虑汇率问题。

    + BIS把它放进了互换

12. 商品互换

##### 回购交易
+ $t=0$：$ 正回购方 {}^{债券}\rightleftarrows_{资金Y} 逆回购方 $
+ $t=T$：$ 正回购方 {}^{债券}\rightleftarrows_{资金Y'} 逆回购方 $
+ 显然：$Y'>Y$,相当于支付了回购利息
+ 分类：
    + 质押式回购：债券所有权不发生转移
    + 买断式回购：债券所有权发生转移
        + 提供了给逆回购方做空市场的机会

+ Fr007：银行间市场7天质押式回购利率，是国内常见的浮动利率

##### 国际互换市场发展迅速的原因
+ 互换交易在风险管理. 降低交易成本. 规避管制和创造新产品等方面都有着重要的运用。

+ 在其发展过程中，互换市场形成的一些*运作机制*也在很大程度上促进了该市场的发展。
    + 做市商制度
    + 合约的标准化
    + 利率互换的其他市场惯例

+ 当局的*监管态度*为互换交易提供了合法发展的空间

#### 互换的运作机制

##### 1. 做市商制度（Market Maker）
+ 互换：OTC产品

+ 经纪制度－>做市商制度

    + 互换市场的做市商被称为互换交易商或互换银行（Swap Bank）

+ 利率互换市场的做市商制度尤其发达：

    + 利率互换的同质性较强

    + 利率风险的套保比较容易

        + 利率的衍生品市场比较发达

##### 2. 合约的标准化

    + International Swaps Dealers Association（最早只做互换） → International Swaps and Derivatives Association（后来扩展至其它衍生品）
    
+ ISDA主协议主要包括
    + 协议主文

+ 利率互换的定价，等价于计算债券组合的价值，也等价于计算 FRA组合的价值。

    + 由于都是列（4）现金流的不同分解，在不考虑不同产品的信用风险和流动性风险差异的情况下，这两种定价结果必然是等价的。

+ 与远期合约相似，利率互换的定价有两种情形：
    1. 在协议签订后的互换定价，是根据协议内容与市场利率水平确定利率互换合约的价值，可能为正，也可能为负。

    2. 在协议签订时，一个公平的利率互换协议应使得双方的互换价值相等。因此协议签订时的互换定价，就是选择一个使得互换的初始价值为零的互换利率
        + t=0时，互换合约价值为0

##### 债券组合定价法
+ 互换多头：$V_{long} = B_{float} - B_{fix} $

+ 互换空头：$V_{long} = -B_{float} + B_{fix} $

+ 已知债券定价方法：

    + $B_{fix} = \Sigma_{i=1}^{n}CF_{i}* e^{-r_{i}*t_{i}} $
        + $r_{i}$ 为每一次计息时刻 $t_{i}$ 的到期收益率，此时的YTM由浮动利率决定

    + $B_{float} = (A+k^{*})*e^{-r_{1}t_{1}} $
        + $ k^{*} $ 为下一次付息的利息，$r_{1}$ 为这一期期初的浮动利率，*贴现率等于浮动利率*
        + 选定证券存续期内的任一时点， 证券的价值等于该时刻的证券价值加上现在到该时点之间现金流的贴现值
            + 因为只能知道下一次付息的利息，后面的信息都不知道 
        + 在每一期重定利率的时间点（利息已经付完），债券价格正好等于面值
        + 在发行时间点(t=0)，债券价格也等于面值

        


##### FRA定价法
+ $V_{long} = A(e^{r_{F}(T^{*}-T)}-e^{r_{K}(T^{*}-T)})e^{-r^{*}(T^{*}-T)} $
    + $ V_{short} = -V_{long} $
    + 从利率期限结构中估计出FRA对应的远期利率，即可得到每笔FRA的价值，加总即为利率互换多头的价值
        + 对每一期现金互换求FRA价值，然后求和
    + $r_{F}$ 是由即期浮动利率估计的远期利率，$r_{K}$ 为固定利率，$r^{*}$ 为贴现率（等于即期浮动利率）
        + $ r_{F} = \frac{[r^{*}(T^{*}-t) - r(T-t)]}{(T^{*}-T)}$

##### 确定合理的互换利率（固定利率）
+ 合理的互换利率就是使得*利率互换价值为零*的固定利率
    + 我们所求的固定利率是*息票率*而非贴现率
    + 债券组合法：$B_{fix} = B_{float}$
        + 若浮动利率债券是新发行的，则 $B_{fix} = B_{float} = A_{B_{float}} $，固定利率债券价值等于浮动利率债券的本金

##### 互换利率
+ 利率互换协议中合理的固定利率就是使得互换价值为零的利率水平，也就是通常所说的互换利率。
+ 现实中的互换利率是市场以一定的计息频率为基础、就特定期限形成的互换中间利率。

##### 互换收益率曲线
+ 固定利率债券的票面利率就是互换利率，浮动利率债券的利率就是SHIBOR。因此，互换利率与SHIBOR存在内在的联系
    + 加上所覆盖的时间期限(隔夜~1年)不同，SHIBOR与互换利率常常互相补充，形成市场中所称的“互换利率期限结构”或“互换收益率曲线”
+ 一般来说，1年以下的短期期限使用SHIBOR，中长期期限则使用*互换利率中隐含的即期利率*
    + 因为此时互换中的固定利率债券不一定是平价债券，故互换利率（息票率）不一定等于零息债券的到期收益率
    + 互换利率实际上是附息票债券的平价到期收益率，并非即期利率，故此在构造互换利率期限结构时，需要从平价到期收益率信息中提取出相应期限的即期利率
    + 所谓平价到期收益率，是指在当前即期利率期限结构下，使得附息票债券价格等于面值的票面利率
        + 对于0~T期的互换而言：$ CF_{0}e^{-r_{0}*0}+CF_{0}e^{-r_{1}*1}+...+CF_{T}e^{-r_{T}*T} = A_{B_{float}} $,通过解出 $r_{T}$,即可得到此时的即期浮动利率
    


+ 互换曲线不是无风险的利率期限结构，但与国债利率期限结构相比，互换曲线有其自身的优势：
    + 新的互换*会在市场上不断产生*，这使得特定期限的互换利率具有*延续性*，几乎每天都可以估计出特定期限的互换利率；而特定期限的国债利率则往往只有在国债新发行之后才能更像，以发行期限进行循环。
    + 互换是零成本合约，其*供给是无限*的，不受发行量的制约和影响；而国债由于供给的制约会产生新券（on the run）和旧券（off the run）的价格差异。
    + 对于许多银行间的金融衍生产品来说，与无风险利率相比，互换利率由于反映了其现金流的*信用风险*与*流动性风险*，是一个更好的贴现率基准

+ 无风险利率期限结构是根据国债价格信息构造的，互换收益率曲线则主要基于银行间市场报价形成的，这两者的差异被称为“互换价差（互换利差）”（swap spread），主要体现了*银行间市场的信用风险、流动性风险与税收*等的差异
    + 显然，互换收益率高于国债收益率

#### 货币互换的定价

+ 货币互换的现金流
    1. 期初交换本金
    2. 期间交换利息
    3. 期末交换本金

##### 运用债券组合为货币互换定价

+ 对于收入本币利息、付出外币利息的那一方：
    + $V^{D-long}_{swap} = B_{D}-S_{0}B_{F} $
        + $S_{0}$ 为当前即期汇率（直接标价：N外币/本币）
        + $B_{D}$ 为本币债券，$B_{F}$ 为外币债券
+ 对于付出本币利息、收入外币利息的那一方:
    + $V^{D-short}_{swap} = -V^{long}_{swap} = S_{0}B_{F} - B_{D} $

##### 运用FXA组合为货币互换定价

+ 计算并加总货币互换中分解出来的每笔FXA的价值，就可得到相应货币互换的价值
+ 对于每一笔FXA:$ f = (A*F-D)e^{-r(T-t)} $
    + A为外币利息；D为本币利息。
    + 此时是收外币利息，付本币利息

#### 互换的风险

+ 与互换相联系的风险主要包括：
    + 信用风险
        + OTC交易的信用风险（对手方风险）
            + 当利率或汇率等市场价格的变动使得互换对交易者而言价值为正时存在
        + 利率互换
            + 交换的仅是*利息差额*，其交易双方真正面临的信用风险暴露远比互换的名义本金要少得多
        + 货币互换
            + 由于进行*本金*的交换，其交易双方面临的信用风险比利率互换要大一些
    + 市场风险
        + 与互换相联系的市场风险主要可分为利率风险和汇率风险：
            + 对于利率互换来说，主要的市场风险是利率风险
            + 对于货币互换而言，市场风险包括利率风险和汇率风险
        + 利率风险的管理：
            + 久期、凸性等分析工具，运用市场上的固定收益产品如欧洲美元期货等对冲
        + 汇率风险的管理：
            + 远期外汇协议等
+ 两者相互影响，相互作用

+ 互换交易中的信用风险是很难估计的，交易者通常通过信用增强（ Credit Enhancement）来管理和消除信用风险。
    + 净额结算
        + 减小风险敞口
    + 抵押和盯市
        + 当价值为负时，要求增加抵押
    + 信用衍生产品
        + CDS
