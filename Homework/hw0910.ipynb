4. 投资者A于2019年11月13日按每股0.0500元的价格卖出一份11月到期行权价为 2.70 的 50ETF 认沽期权，共收入期权费 500 元。该期权合约的前结算价为0.048元，50ETF前收盘价为2.49元。 
   
   1. 请计算该投资者应缴纳的开仓保证金。 
   
   2. 11月13日，50ETF收盘价为2.47元，上述期权的结算价为0.05元，请计算该投资者的维持保证金。 
   
   3. 投资者当天浮动盈亏金额是多少？ 
   
   4. 若开仓时该投资者账户上的资金刚好只够缴纳开仓保证金，11月13日收盘后，该投资者是否需要补缴保证金？（请计算说明）

1. \begin{aligned}
   认沽期权义务仓开仓保证金 &＝Min[合约前结算价+Max(12\%×合约标的前收盘价-认沽期权虚值,\\
   & 7\%×行权价格),行权价格]×合约单位\\
   &= Min[0.048+Max(0.12*2.49,0.07*2.7),2.7]*\frac{500}{0.05}\\
   &= 3468 元
   \end{aligned}

2. \begin{aligned}
   认沽期权义务仓维持保证金&＝Min[合约结算价 +Max(12\%×合约标的收盘价-认沽期权虚值,\\
   & 7\%×行权价格),行权价格]×合约单位\\
   &= Min[0.05 + Max(0.12* 2.47,0.07 * 2.7),2.7 ]* \frac{500}{0.05}\\
   &= 3464 元
   \end{aligned}

3. $盈亏 =  (0.05-0.05)*\frac{500}{0.05} = 0$

4. $保证金缺口 = 维持保证金 - 开仓保证金 = -4 < 0$
   + 因为不存在正的保证金缺口，所以不需要补交保证金


```python
init_margin = 500/0.05*min(0.048+max(0.12*2.49,0.07*2.7),2.7)
print(f'开仓保证金:{init_margin}')
main_margin = 500/0.05*min(0.05+max(0.12*2.47,0.07*2.7),2.7)
print(f'开仓保证金:{main_margin}')
profit = 500/0.05*(0.05-0.05) 
print(f'盈亏：{profit}')
```

5. 某支付红利的股票现在价格为 30.00 元，3 个月后将支付 1 元现金红利，该股票行权价为28.00元的6个月期欧式看涨期权的期权费为5.32元，具有相同行权价和相同期限的该股票欧式看跌期权的期权费为3.57元。假定无风险存款的连续复利利率为 3.5%。
   
   + 请问存在套利机会吗？
   
   + 若存在如何操作？请详细叙述每个时刻的现金流量

+ 由题意得，$S_{t} = 30$,$r= 3.5\%$,$I = 1*e^{-r*\frac{1}{4}} = 0.9912881698401698$,$K=28$,$p^{E}_{t} = 3.57$，$c^{E}_{t} = 5.32$,$T-t = \frac{1}{2}$
+ 无套利情况下，有看跌看涨平价公式：$$ c^{E}_{t} + K*e^{-r(T-t)} = p^{E}_{t} + S_{t} + I $$
+ 因为$ c^{E}_{t} + K*e^{-r(T-t)} = 32.83426259862205 < 34.56128816984017 = p^{E}_{t} + S_{t} + I $，因此存在套利机会,套利策略如下：

\begin{array}{c|c|c|c}
    \text{套利策略} & t\text{时刻} & T\text{时刻}(S_{K} + I*e^{r(T-t)}>K ) & T\text{时刻}(S_{K} + I*e^{r(T-t)}\leq K ) \\
    \hline
    +c^{E}_{t} & -5.32 & (S_{T}-K) & 0    \\
    \hline
    -p^{E}_{t}  & +3.57 & 0 & -(K-S_{T})    \\
    \hline    
    +K*e^{-r(T-t)} & -27.51426259862205 & K & K \\
    \hline
    -S_{t} & +30.99128816984017 & -S_{T} & -S_{T} \\
    \hline
    Total \space return & 1.727025571218121 & 0 & 0 \\
\end{array}

```python
from math import e
I = 1*e**(-0.035*0.25)
short = 3.57 + 30 + I
long = 5.32 + 28*e**(-0.035*0.5)
payoff = short-long
print(f'套利获益:{payoff}')
```
